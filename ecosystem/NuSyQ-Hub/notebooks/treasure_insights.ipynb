```python
# Imports and data load
from pathlib import Path
import json, math
AGG_JSON = Path('docs/Reports/aggregated_insights.json')
if not AGG_JSON.exists():
    raise FileNotFoundError('Run: python src/tools/report_aggregator.py --json first')
insights = json.loads(AGG_JSON.read_text(encoding='utf-8'))
insights.keys()
```

```python
# Severity distributions (action plan vs issue stubs)
action_counts = insights['action_plan_counts']
issue_counts = insights['issue_stub_counts']
def percent(count, total):
    return 0 if total == 0 else round((count/total)*100, 2)
total_action = sum(action_counts.values())
total_issue = sum(issue_counts.values())
dist_table = []
for sev in ['critical','high','medium','low','info']:
    dist_table.append((sev, action_counts.get(sev,0), percent(action_counts.get(sev,0), total_action), issue_counts.get(sev,0), percent(issue_counts.get(sev,0), total_issue)))
dist_table
```

```python
# Textual bar visualization (no external libs for offline-first)
def bar(count, max_count, width=40):
    if max_count == 0: return ''
    filled = math.ceil((count/max_count)*width)
    return '#' * filled + '-' * (width - filled)
max_action = max(action_counts.values()) if action_counts else 0
max_issue = max(issue_counts.values()) if issue_counts else 0
print('Action Plan Severity Distribution:')
for sev, cnt, pct_a, _, _ in dist_table:
    print(f'{sev.title():9} {cnt:4} ({pct_a:5}%) | {bar(cnt, max_action)}')
print('')
print('Issue Stub Severity Distribution:')
for sev, _, _, cnt_i, pct_i in dist_table:
    print(f'{sev.title():9} {cnt_i:4} ({pct_i:5}%) | {bar(cnt_i, max_issue)}')
```
