## 2️⃣ Load Capability Inventory and Tool Registry

## 5️⃣ Extension Configuration Audit  
Checking for Jupyter, Obsidian, ChatDev, and other VSCode extension configurations

```python
# Check VSCode extension configuration
extensions_path = repo_root / ".vscode" / "extensions.json"

if extensions_path.exists():
    with open(extensions_path) as f:
        extensions_config = json.load(f)
    
    print("✅ Extensions configuration found")
    print(f"\n📦 Recommended Extensions: {len(extensions_config.get('recommendations', []))}")
    
    # Check for key integrations
    recommendations = extensions_config.get("recommendations", [])
    
    key_extensions = {
        "Jupyter": ["ms-toolsai.jupyter", "ms-toolsai.jupyter-keymap", "ms-toolsai.jupyter-renderers"],
        "Python": ["ms-python.python", "ms-python.vscode-pylance"],
        "GitHub Copilot": ["github.copilot", "github.copilot-chat"],
        "Continue.dev (Ollama)": ["continue.continue"],
        "Git": ["mhutchie.git-graph", "eamodio.gitlens"],
        "Markdown": ["yzhang.markdown-all-in-one", "bierner.markdown-mermaid"]
    }
    
    print("\n✅ Key Extension Status:")
    for category, ext_list in key_extensions.items():
        found = [ext for ext in ext_list if ext in recommendations]
        status = "✅" if found else "❌"
        print(f"   {status} {category}: {len(found)}/{len(ext_list)} installed")
        if found:
            for ext in found:
                print(f"      └─ {ext}")
    
    # Check for custom integrations
    print("\n🔧 Custom Integrations Needed:")
    custom_integ = extensions_config.get("custom_integrations_needed", {})
    for name, config in custom_integ.items():
        status = config.get("status", "unknown")
        print(f"   ├─ {name}: {status}")
        print(f"   │  └─ {config.get('description', 'No description')}")
        
else:
    print(f"❌ Extensions configuration not found at: {extensions_path}")
```
