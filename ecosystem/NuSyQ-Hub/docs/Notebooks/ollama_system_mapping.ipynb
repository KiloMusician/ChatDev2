## 🔍 Section 1: Scan System for Ollama Processes

This section identifies all running Ollama processes, checks service status, and analyzes resource usage.

````python
def scan_ollama_processes():
    """Comprehensive scan for Ollama processes and services"""
    print("🔍 Scanning for Ollama processes...")
    
    ollama_processes = {
        "running_processes": [],
        "service_status": "unknown",
        "api_accessibility": False,
        "models_available": [],
        "resource_usage": {},
        "connection_details": {}
    }
    
    # Method 1: Use ProcessManager if available
    if PROCESS_MANAGER_AVAILABLE:
        try:
            process_manager = ProcessManager()
            system_processes = process_manager.scan_existing_processes()
            
            # Look for Ollama in system processes
            for proc_list in system_processes.values():
                for proc in proc_list:
                    if "ollama" in proc.get("name", "").lower() or \
                       any("ollama" in str(arg).lower() for arg in proc.get("cmdline", [])):
                        ollama_processes["running_processes"].append(proc)
            
            print(f"✅ ProcessManager scan: {len(ollama_processes['running_processes'])} Ollama processes found")
        except Exception as e:
            print(f"⚠️ ProcessManager scan failed: {e}")
    
    # Method 2: Direct psutil scan
    if PSUTIL_AVAILABLE:
        try:
            for proc in psutil.process_iter(["pid", "name", "cmdline", "cpu_percent", "memory_info", "create_time"]):
                try:
                    proc_info = proc.info
                    proc_name = proc_info.get("name", "").lower()
                    cmdline = " ".join(proc_info.get("cmdline", [])).lower()
                    
                    if "ollama" in proc_name or "ollama" in cmdline:
                        process_data = {
                            "pid": proc_info["pid"],
                            "name": proc_info["name"],
                            "cmdline": proc_info["cmdline"],
                            "cpu_percent": proc_info.get("cpu_percent", 0),
                            "memory_mb": proc_info.get("memory_info", {}).get("rss", 0) / (1024*1024) if proc_info.get("memory_info") else 0,
                            "create_time": datetime.fromtimestamp(proc_info.get("create_time", 0)).isoformat(),
                            "status": proc.status()
                        }
                        ollama_processes["running_processes"].append(process_data)
                        
                except (psutil.NoSuchProcess, psutil.AccessDenied):
                    continue
                    
            print(f"✅ PSUtil scan: {len(ollama_processes['running_processes'])} Ollama processes found")
        except Exception as e:
            print(f"⚠️ PSUtil scan failed: {e}")
    
    # Method 3: Command line check
    try:
        result = subprocess.run(["ollama", "ps"], capture_output=True, text=True, timeout=10)
        if result.returncode == 0:
            ollama_processes["service_status"] = "running"
            print("✅ Ollama CLI accessible - service is running")
        else:
            ollama_processes["service_status"] = "not_running"
            print("❌ Ollama CLI not accessible")
    except FileNotFoundError:
        print("❌ Ollama CLI not found in PATH")
        ollama_processes["service_status"] = "not_installed"
    except Exception as e:
        print(f"⚠️ Ollama CLI check failed: {e}")
    
    # Method 4: API connectivity test
    try:
        response = requests.get("http://localhost:11434/api/tags", timeout=5)
        if response.status_code == 200:
            ollama_processes["api_accessibility"] = True
            models_data = response.json()
            ollama_processes["models_available"] = models_data.get("models", [])
            ollama_processes["connection_details"] = {
                "endpoint": "http://localhost:11434",
                "response_time": response.elapsed.total_seconds(),
                "api_version": response.headers.get("server", "unknown")
            }
            print(f"✅ Ollama API accessible - {len(ollama_processes['models_available'])} models available")
        else:
            print(f"❌ Ollama API returned status: {response.status_code}")
    except requests.exceptions.ConnectionError:
        print("❌ Ollama API not accessible (connection refused)")
    except Exception as e:
        print(f"⚠️ Ollama API check failed: {e}")
    
    # Store results
    ollama_analysis["processes"] = ollama_processes
    
    return ollama_processes

# Execute the scan
process_results = scan_ollama_processes()

# Display summary
print("\n📊 OLLAMA PROCESS SCAN SUMMARY")
print("=" * 50)
print(f"🔄 Running Processes: {len(process_results['running_processes'])}")
print(f"🔧 Service Status: {process_results['service_status']}")
print(f"🌐 API Accessible: {process_results['api_accessibility']}")
print(f"📦 Models Available: {len(process_results['models_available'])}")

if process_results["running_processes"]:
    print("\n🔍 Process Details:")
    for proc in process_results["running_processes"]:
        print(f"  PID {proc.get('pid', 'unknown')}: {proc.get('name', 'unknown')} "
              f"(CPU: {proc.get('cpu_percent', 0):.1f}%, "
              f"RAM: {proc.get('memory_mb', 0):.1f}MB)")

if process_results["models_available"]:
    print("\n📦 Available Models:")
    for model in process_results["models_available"][:5]:  # First 5 models
        name = model.get("name", "unknown")
        size_gb = model.get("size", 0) / (1024**3)
        print(f"  📋 {name} ({size_gb:.1f} GB)")
    if len(process_results["models_available"]) > 5:
        print(f"  ... and {len(process_results['models_available']) - 5} more models")
````

## 🔎 Section 2: Search Codebase for Ollama References

This section recursively searches through all repository files to identify Ollama references, imports, API calls, and related code.

````python
def search_ollama_references():
    """Comprehensive search for Ollama references in codebase"""
    print("🔎 Searching codebase for Ollama references...")
    
    ollama_references = {
        "python_files": [],
        "config_files": [],
        "documentation": [],
        "scripts": [],
        "import_statements": [],
        "api_calls": [],
        "configuration_references": [],
        "total_files": 0,
        "total_references": 0
    }
    
    # Search patterns
    search_patterns = {
        "import_patterns": [
            r"import\s+ollama",
            r"from\s+ollama",
            r"from\s+.*ollama",
            r"import\s+.*ollama.*"
        ],
        "api_patterns": [
            r"localhost:11434",
            r"11434",
            r"/api/generate",
            r"/api/tags", 
            r"/api/chat",
            r"ollama\.generate",
            r"ollama\.chat",
            r"ollama\.Client"
        ],
        "reference_patterns": [
            r"ollama",
            r"Ollama",
            r"OLLAMA"
        ]
    }
    
    # File extensions to search
    searchable_extensions = {".py", ".json", ".yaml", ".yml", ".md", ".txt", ".ps1", ".sh"}
    skip_dirs = {".git", "__pycache__", ".vscode", "node_modules", ".pytest_cache", "venv", "env"}
    
    def search_file_content(file_path: Path, content: str) -> Dict[str, Any]:
        """Search file content for Ollama patterns"""
        file_results = {
            "path": str(file_path.relative_to(repo_root)),
            "type": file_path.suffix,
            "size": file_path.stat().st_size,
            "imports": [],
            "api_calls": [],
            "references": [],
            "line_matches": []
        }
        
        lines = content.split("\n")
        
        for line_num, line in enumerate(lines, 1):
            line_lower = line.lower()
            
            # Check for imports
            for pattern in search_patterns["import_patterns"]:
                if re.search(pattern, line, re.IGNORECASE):
                    file_results["imports"].append({
                        "line": line_num,
                        "content": line.strip(),
                        "pattern": pattern
                    })
            
            # Check for API calls
            for pattern in search_patterns["api_patterns"]:
                if re.search(pattern, line, re.IGNORECASE):
                    file_results["api_calls"].append({
                        "line": line_num,
                        "content": line.strip(),
                        "pattern": pattern
                    })
            
            # Check for general references
            if "ollama" in line_lower:
                file_results["references"].append({
                    "line": line_num,
                    "content": line.strip()
                })
                file_results["line_matches"].append(line_num)
        
        return file_results
    
    # Search through repository
    try:
        for root, dirs, files in os.walk(repo_root):
            # Skip certain directories
            dirs[:] = [d for d in dirs if d not in skip_dirs]
            
            for file in files:
                file_path = Path(root) / file
                
                # Check if file extension is searchable
                if file_path.suffix not in searchable_extensions:
                    continue
                
                try:
                    with open(file_path, "r", encoding="utf-8", errors="ignore") as f:
                        content = f.read()
                    
                    # Search file content
                    file_results = search_file_content(file_path, content)
                    
                    # Only include files with matches
                    total_matches = (len(file_results["imports"]) + 
                                   len(file_results["api_calls"]) + 
                                   len(file_results["references"]))
                    
                    if total_matches > 0:
                        ollama_references["total_references"] += total_matches
                        
                        # Categorize by file type
                        if file_path.suffix == ".py":
                            ollama_references["python_files"].append(file_results)
                        elif file_path.suffix in {".json", ".yaml", ".yml"}:
                            ollama_references["config_files"].append(file_results)
                        elif file_path.suffix == ".md":
                            ollama_references["documentation"].append(file_results)
                        elif file_path.suffix in {".ps1", ".sh"}:
                            ollama_references["scripts"].append(file_results)
                        
                        # Collect specific patterns
                        ollama_references["import_statements"].extend(file_results["imports"])
                        ollama_references["api_calls"].extend(file_results["api_calls"])
                        
                        ollama_references["total_files"] += 1
                
                except (PermissionError, UnicodeDecodeError):
                    continue
        
        print(f"✅ Search complete: {ollama_references['total_files']} files with {ollama_references['total_references']} references")
        
    except Exception as e:
        print(f"❌ Search failed: {e}")
    
    # Store results
    ollama_analysis["file_references"] = ollama_references
    
    return ollama_references

# Execute the search
reference_results = search_ollama_references()

# Display detailed results
print("\n📊 OLLAMA REFERENCE SEARCH RESULTS")
print("=" * 50)
print(f"📁 Total Files with References: {reference_results['total_files']}")
print(f"🔍 Total References Found: {reference_results['total_references']}")
print(f"🐍 Python Files: {len(reference_results['python_files'])}")
print(f"⚙️ Config Files: {len(reference_results['config_files'])}")
print(f"📝 Documentation: {len(reference_results['documentation'])}")
print(f"📜 Scripts: {len(reference_results['scripts'])}")

# Show top Python files with most references
if reference_results["python_files"]:
    print("\n🐍 Top Python Files by Reference Count:")
    python_sorted = sorted(reference_results["python_files"], 
                          key=lambda x: len(x["imports"]) + len(x["api_calls"]) + len(x["references"]), 
                          reverse=True)
    
    for file_data in python_sorted[:10]:  # Top 10
        total_refs = len(file_data["imports"]) + len(file_data["api_calls"]) + len(file_data["references"])
        print(f"  📄 {file_data['path']} ({total_refs} references)")
        if file_data["imports"]:
            print(f"    📥 {len(file_data['imports'])} imports")
        if file_data["api_calls"]:
            print(f"    🌐 {len(file_data['api_calls'])} API calls")

# Show import analysis
if reference_results["import_statements"]:
    print(f"\n📥 Import Statement Analysis ({len(reference_results['import_statements'])} total):")
    import_types = defaultdict(int)
    for imp in reference_results["import_statements"]:
        import_types[imp["pattern"]] += 1
    
    for pattern, count in sorted(import_types.items(), key=lambda x: x[1], reverse=True):
        print(f"  📋 {pattern}: {count} occurrences")

# Show API call analysis
if reference_results["api_calls"]:
    print(f"\n🌐 API Call Analysis ({len(reference_results['api_calls'])} total):")
    api_types = defaultdict(int)
    for api in reference_results["api_calls"]:
        api_types[api["pattern"]] += 1
    
    for pattern, count in sorted(api_types.items(), key=lambda x: x[1], reverse=True):
        print(f"  🔗 {pattern}: {count} occurrences")
````

## ⚙️ Section 3: Configuration & API Endpoint Analysis

````python
def analyze_ollama_configuration():
    """Analyze Ollama configuration files and API endpoints"""
    print("⚙️ Analyzing Ollama configuration and API endpoints...")
    
    config_analysis = {
        "configuration_files": [],
        "api_endpoints": [],
        "connection_details": {},
        "model_configurations": [],
        "service_definitions": [],
        "environment_variables": [],
        "integration_configs": []
    }
    
    # Configuration file patterns to search for
    
    # API endpoint patterns
    api_patterns = {
        "base_url": r"http://localhost:11434|https?://[^/]*:11434",
        "generate": r"/api/generate",
        "chat": r"/api/chat", 
        "tags": r"/api/tags",
        "models": r"/api/models",
        "pull": r"/api/pull",
        "push": r"/api/push",
        "embeddings": r"/api/embeddings",
        "version": r"/api/version"
    }
    
    def analyze_config_file(file_path: Path) -> Dict[str, Any]:
        """Analyze individual configuration file"""
        try:
            content = file_path.read_text(encoding="utf-8", errors="ignore")
            
            file_analysis = {
                "path": str(file_path.relative_to(repo_root)),
                "type": file_path.suffix,
                "size": file_path.stat().st_size,
                "ollama_keys": [],
                "api_endpoints": [],
                "model_references": [],
                "connection_configs": [],
                "environment_vars": []
            }
            
            # Parse based on file type
            if file_path.suffix in [".yaml", ".yml"]:
                try:
                    import yaml
                    data = yaml.safe_load(content)
                    file_analysis.update(analyze_yaml_config(data, content))
                except:
                    file_analysis.update(analyze_text_config(content))
            
            elif file_path.suffix == ".json":
                try:
                    data = json.loads(content)
                    file_analysis.update(analyze_json_config(data, content))
                except:
                    file_analysis.update(analyze_text_config(content))
            
            else:
                file_analysis.update(analyze_text_config(content))
            
            return file_analysis
            
        except Exception as e:
            return {"path": str(file_path), "error": str(e)}
    
    def analyze_yaml_config(data: Any, content: str) -> Dict[str, Any]:
        """Analyze YAML configuration data"""
        analysis = {"yaml_data": {}}
        
        def search_dict(obj, path=""):
            if isinstance(obj, dict):
                for key, value in obj.items():
                    current_path = f"{path}.{key}" if path else key
                    if "ollama" in key.lower():
                        analysis["ollama_keys"].append({
                            "path": current_path,
                            "key": key,
                            "value": value
                        })
                    search_dict(value, current_path)
            elif isinstance(obj, list):
                for i, item in enumerate(obj):
                    search_dict(item, f"{path}[{i}]")
        
        if data:
            search_dict(data)
            analysis["yaml_data"] = data
        
        # Also search content as text
        analysis.update(analyze_text_config(content))
        return analysis
    
    def analyze_json_config(data: Any, content: str) -> Dict[str, Any]:
        """Analyze JSON configuration data"""
        analysis = {"json_data": {}}
        
        def search_json(obj, path=""):
            if isinstance(obj, dict):
                for key, value in obj.items():
                    current_path = f"{path}.{key}" if path else key
                    if "ollama" in key.lower():
                        analysis["ollama_keys"].append({
                            "path": current_path,
                            "key": key,
                            "value": value
                        })
                    search_json(value, current_path)
            elif isinstance(obj, list):
                for i, item in enumerate(obj):
                    search_json(item, f"{path}[{i}]")
        
        if data:
            search_json(data)
            analysis["json_data"] = data
        
        # Also search content as text
        analysis.update(analyze_text_config(content))
        return analysis
    
    def analyze_text_config(content: str) -> Dict[str, Any]:
        """Analyze configuration as text"""
        analysis = {
            "ollama_keys": [],
            "api_endpoints": [],
            "model_references": [],
            "connection_configs": [],
            "environment_vars": []
        }
        
        lines = content.split("\n")
        
        for line_num, line in enumerate(lines, 1):
            line_clean = line.strip()
            
            # Search for API endpoints
            for endpoint_name, pattern in api_patterns.items():
                if re.search(pattern, line, re.IGNORECASE):
                    analysis["api_endpoints"].append({
                        "line": line_num,
                        "endpoint": endpoint_name,
                        "content": line_clean,
                        "pattern": pattern
                    })
            
            # Search for model references
            model_patterns = [r"llama", r"mistral", r"codellama", r"phi", r"gemma", r"qwen"]
            for pattern in model_patterns:
                if re.search(pattern, line, re.IGNORECASE):
                    analysis["model_references"].append({
                        "line": line_num,
                        "model": pattern,
                        "content": line_clean
                    })
            
            # Search for connection configs
            connection_patterns = [r"host.*11434", r"port.*11434", r"base_url", r"endpoint"]
            for pattern in connection_patterns:
                if re.search(pattern, line, re.IGNORECASE):
                    analysis["connection_configs"].append({
                        "line": line_num,
                        "type": pattern,
                        "content": line_clean
                    })
            
            # Search for environment variables
            env_patterns = [r"OLLAMA_", r"ollama", r"MODEL_"]
            for pattern in env_patterns:
                if re.search(pattern, line, re.IGNORECASE) and ("=" in line or "export" in line.lower()):
                    analysis["environment_vars"].append({
                        "line": line_num,
                        "var": pattern,
                        "content": line_clean
                    })
        
        return analysis
    
    # Search for configuration files 
    config_file_extensions = {".yaml", ".yml", ".json", ".env", ".ps1", ".ini", ".toml"}
    
    try:
        for root, dirs, files in os.walk(repo_root):
            dirs[:] = [d for d in dirs if d not in {".git", "__pycache__", ".vscode", "node_modules"}]
            
            for file in files:
                file_path = Path(root) / file
                
                # Check if it's a configuration file type
                if file_path.suffix not in config_file_extensions:
                    continue
                
                # Check if filename suggests Ollama/AI configuration
                filename_lower = file_path.name.lower()
                is_relevant = (
                    "ollama" in filename_lower or
                    "chatdev" in filename_lower or
                    "ai" in filename_lower or
                    "model" in filename_lower or
                    "config" in filename_lower or
                    "environment" in filename_lower or
                    "settings" in filename_lower
                )
                
                if is_relevant:
                    file_analysis = analyze_config_file(file_path)
                    if "error" not in file_analysis:
                        # Check if file has Ollama-related content
                        has_content = (
                            file_analysis.get("ollama_keys") or
                            file_analysis.get("api_endpoints") or
                            file_analysis.get("model_references") or
                            file_analysis.get("connection_configs") or
                            "ollama" in file_path.read_text(encoding="utf-8", errors="ignore").lower()
                        )
                        
                        if has_content:
                            config_analysis["configuration_files"].append(file_analysis)
                            
                            # Aggregate endpoint data
                            config_analysis["api_endpoints"].extend(file_analysis.get("api_endpoints", []))
                            config_analysis["model_configurations"].extend(file_analysis.get("model_references", []))
                            config_analysis["environment_variables"].extend(file_analysis.get("environment_vars", []))
    
    except Exception as e:
        print(f"❌ Configuration analysis failed: {e}")
    
    # Test API connectivity
    config_analysis["connection_details"] = test_ollama_connectivity()
    
    # Store results
    ollama_analysis["configuration"] = config_analysis
    
    return config_analysis

def test_ollama_connectivity():
    """Test connection to Ollama API"""
    print("  🔗 Testing Ollama API connectivity...")
    
    connection_results = {
        "base_url": "http://localhost:11434",
        "status": "unknown",
        "version": None,
        "models": [],
        "endpoints_tested": {},
        "response_times": {}
    }
    
    try:
        from datetime import datetime

        import requests
        
        base_url = "http://localhost:11434"
        timeout = 5
        
        # Test endpoints
        endpoints_to_test = {
            "version": "/api/version",
            "tags": "/api/tags",
            "generate": "/api/generate"
        }
        
        for name, endpoint in endpoints_to_test.items():
            try:
                start_time = datetime.now()
                
                if name == "generate":
                    # Test with minimal generate request
                    response = requests.post(
                        f"{base_url}{endpoint}",
                        json={"model": "test", "prompt": "test", "stream": False},
                        timeout=timeout
                    )
                else:
                    response = requests.get(f"{base_url}{endpoint}", timeout=timeout)
                
                end_time = datetime.now()
                response_time = (end_time - start_time).total_seconds()
                
                connection_results["endpoints_tested"][name] = {
                    "status_code": response.status_code,
                    "accessible": response.status_code < 500,
                    "response_time": response_time
                }
                connection_results["response_times"][name] = response_time
                
                # Parse useful data
                if name == "version" and response.status_code == 200:
                    try:
                        connection_results["version"] = response.json()
                    except:
                        connection_results["version"] = response.text
                
                elif name == "tags" and response.status_code == 200:
                    try:
                        tags_data = response.json()
                        connection_results["models"] = [model.get("name", "unknown") for model in tags_data.get("models", [])]
                    except:
                        pass
                
            except requests.exceptions.ConnectionError:
                connection_results["endpoints_tested"][name] = {
                    "status_code": None,
                    "accessible": False,
                    "error": "Connection refused"
                }
            except requests.exceptions.Timeout:
                connection_results["endpoints_tested"][name] = {
                    "status_code": None,
                    "accessible": False,
                    "error": "Timeout"
                }
            except Exception as e:
                connection_results["endpoints_tested"][name] = {
                    "status_code": None,
                    "accessible": False,
                    "error": str(e)
                }
        
        # Determine overall status
        accessible_endpoints = sum(1 for endpoint in connection_results["endpoints_tested"].values() if endpoint["accessible"])
        if accessible_endpoints > 0:
            connection_results["status"] = "connected"
            print(f"  ✅ Ollama API accessible ({accessible_endpoints}/{len(endpoints_to_test)} endpoints)")
        else:
            connection_results["status"] = "disconnected"
            print("  ❌ Ollama API not accessible")
    
    except ImportError:
        connection_results["status"] = "no_requests"
        print("  ⚠️ requests library not available for connectivity test")
    except Exception as e:
        connection_results["status"] = "error"
        connection_results["error"] = str(e)
        print(f"  ❌ Connectivity test failed: {e}")
    
    return connection_results

# Execute configuration analysis
config_results = analyze_ollama_configuration()

# Display results
print("\n⚙️ OLLAMA CONFIGURATION ANALYSIS")
print("=" * 50)
print(f"📁 Configuration Files Found: {len(config_results['configuration_files'])}")
print(f"🔗 API Endpoints Found: {len(config_results['api_endpoints'])}")
print(f"🤖 Model References: {len(config_results['model_configurations'])}")
print(f"🌍 Environment Variables: {len(config_results['environment_variables'])}")

# Show connection status
conn_details = config_results.get("connection_details", {})
print(f"\n🔗 API Connection Status: {conn_details.get('status', 'unknown').upper()}")
if conn_details.get("models"):
    print(f"🤖 Available Models: {', '.join(conn_details['models'][:5])}{'...' if len(conn_details['models']) > 5 else ''}")

# Show configuration files
if config_results["configuration_files"]:
    print("\n📁 Configuration Files with Ollama References:")
    for config_file in config_results["configuration_files"][:10]:  # Top 10
        path = config_file["path"]
        endpoints = len(config_file.get("api_endpoints", []))
        models = len(config_file.get("model_references", []))
        print(f"  📄 {path}")
        if endpoints:
            print(f"    🔗 {endpoints} API endpoints")
        if models:
            print(f"    🤖 {models} model references")

# Show endpoint analysis
if config_results["api_endpoints"]:
    print(f"\n🔗 API Endpoint Analysis ({len(config_results['api_endpoints'])} total):")
    endpoint_counts = defaultdict(int)
    for endpoint in config_results["api_endpoints"]:
        endpoint_counts[endpoint["endpoint"]] += 1
    
    for endpoint, count in sorted(endpoint_counts.items(), key=lambda x: x[1], reverse=True):
        print(f"  🌐 {endpoint}: {count} references")
````

## 🔄 Section 4: Integration Mapping & Dependency Analysis

````python
def map_ollama_integrations():
    """Map Ollama integrations and analyze dependencies"""
    print("🔄 Mapping Ollama integrations and dependencies...")
    
    integration_map = {
        "core_modules": [],
        "integration_layers": [],
        "dependency_graph": {},
        "import_relationships": [],
        "class_hierarchies": [],
        "function_calls": [],
        "data_flows": [],
        "integration_patterns": {}
    }
    
    # Key integration files discovered from semantic search
    key_files = [
        "src/integration/ollama_integration.py",
        "src/ai/ollama_chatdev_integrator.py", 
        "src/integration/chatdev_llm_adapter.py",
        "src/integration/Update-ChatDev-to-use-Ollama.py",
        "src/ai/ollama_model_manager.py"
    ]
    
    def analyze_integration_file(file_path: Path) -> Dict[str, Any]:
        """Analyze individual integration file for patterns and dependencies"""
        if not file_path.exists():
            return {"path": str(file_path), "exists": False}
        
        try:
            content = file_path.read_text(encoding="utf-8", errors="ignore")
            
            file_analysis = {
                "path": str(file_path.relative_to(repo_root)),
                "exists": True,
                "size": file_path.stat().st_size,
                "lines": len(content.split("\n")),
                "imports": [],
                "classes": [],
                "functions": [],
                "ollama_methods": [],
                "api_calls": [],
                "dependency_imports": [],
                "integration_patterns": []
            }
            
            lines = content.split("\n")
            current_class = None
            current_function = None
            
            for line_num, line in enumerate(lines, 1):
                line_stripped = line.strip()
                
                # Track imports
                if line_stripped.startswith(("import ", "from ")):
                    file_analysis["imports"].append({
                        "line": line_num,
                        "statement": line_stripped
                    })
                    
                    # Track dependency imports specifically
                    dependency_keywords = ["ollama", "chatdev", "requests", "json", "asyncio", "subprocess"]
                    if any(keyword in line_stripped.lower() for keyword in dependency_keywords):
                        file_analysis["dependency_imports"].append({
                            "line": line_num,
                            "statement": line_stripped,
                            "type": "dependency"
                        })
                
                # Track class definitions
                if line_stripped.startswith("class "):
                    class_match = re.match(r"class\s+(\w+)", line_stripped)
                    if class_match:
                        current_class = class_match.group(1)
                        file_analysis["classes"].append({
                            "line": line_num,
                            "name": current_class,
                            "definition": line_stripped
                        })
                
                # Track function/method definitions
                if line_stripped.startswith("def "):
                    func_match = re.match(r"def\s+(\w+)", line_stripped)
                    if func_match:
                        current_function = func_match.group(1)
                        func_info = {
                            "line": line_num,
                            "name": current_function,
                            "definition": line_stripped,
                            "class": current_class
                        }
                        file_analysis["functions"].append(func_info)
                        
                        # Check if it's an Ollama-specific method
                        if "ollama" in current_function.lower():
                            file_analysis["ollama_methods"].append(func_info)
                
                # Track API calls and Ollama interactions
                api_patterns = [
                    r"requests\.(get|post|put|delete)",
                    r"ollama\.\w+",
                    r"localhost:11434",
                    r"/api/(generate|chat|tags|models)",
                    r"\.generate\(",
                    r"\.chat\(",
                    r"ChatDev",
                    r"LLM"
                ]
                
                for pattern in api_patterns:
                    if re.search(pattern, line, re.IGNORECASE):
                        file_analysis["api_calls"].append({
                            "line": line_num,
                            "content": line_stripped,
                            "pattern": pattern,
                            "function": current_function,
                            "class": current_class
                        })
                
                # Identify integration patterns
                integration_pattern_keywords = [
                    ("adapter", "Adapter Pattern"),
                    ("bridge", "Bridge Pattern"),
                    ("proxy", "Proxy Pattern"),
                    ("factory", "Factory Pattern"),
                    ("singleton", "Singleton Pattern"),
                    ("observer", "Observer Pattern"),
                    ("async", "Async Pattern"),
                    ("callback", "Callback Pattern")
                ]
                
                for keyword, pattern_name in integration_pattern_keywords:
                    if keyword in line_stripped.lower():
                        file_analysis["integration_patterns"].append({
                            "line": line_num,
                            "pattern": pattern_name,
                            "content": line_stripped
                        })
            
            return file_analysis
            
        except Exception as e:
            return {"path": str(file_path), "error": str(e)}
    
    def build_dependency_graph():
        """Build dependency graph from import relationships"""
        graph = defaultdict(set)
        
        for file_analysis in integration_map["core_modules"]:
            file_path = file_analysis["path"]
            
            for imp in file_analysis.get("dependency_imports", []):
                statement = imp["statement"]
                
                # Extract module names from import statements
                if statement.startswith("from "):
                    # from module import something
                    match = re.match(r"from\s+([^\s]+)", statement)
                    if match:
                        module = match.group(1)
                        graph[file_path].add(module)
                elif statement.startswith("import "):
                    # import module
                    match = re.match(r"import\s+([^\s,]+)", statement)
                    if match:
                        module = match.group(1)
                        graph[file_path].add(module)
        
        return dict(graph)
    
    def analyze_data_flows():
        """Analyze data flows between Ollama components"""
        flows = []
        
        # Define common data flow patterns in Ollama integrations
        flow_patterns = [
            ("prompt", "input", "User input to Ollama"),
            ("response", "output", "Ollama response output"),
            ("model", "configuration", "Model selection/config"),
            ("stream", "streaming", "Streaming response handling"),
            ("context", "memory", "Context/conversation memory"),
            ("error", "error_handling", "Error handling flow"),
            ("callback", "async", "Async callback handling")
        ]
        
        for file_analysis in integration_map["core_modules"]:
            file_content = None
            try:
                file_path = repo_root / file_analysis["path"]
                if file_path.exists():
                    file_content = file_path.read_text(encoding="utf-8", errors="ignore")
            except:
                continue
            
            if not file_content:
                continue
            
            for keyword, flow_type, description in flow_patterns:
                if keyword in file_content.lower():
                    flows.append({
                        "file": file_analysis["path"],
                        "flow_type": flow_type,
                        "keyword": keyword,
                        "description": description,
                        "occurrences": file_content.lower().count(keyword)
                    })
        
        return flows
    
    # Analyze key integration files
    print("  📄 Analyzing integration files...")
    for file_path_str in key_files:
        file_path = repo_root / file_path_str
        analysis = analyze_integration_file(file_path)
        if analysis.get("exists"):
            integration_map["core_modules"].append(analysis)
    
    # Also search for additional integration files
    additional_patterns = [
        "**/ollama*.py",
        "**/chatdev*.py", 
        "**/*llm*.py",
        "**/*ai*.py"
    ]
    
    for pattern in additional_patterns:
        for file_path in repo_root.glob(pattern):
            if file_path.is_file() and file_path.suffix == ".py":
                # Avoid duplicates
                relative_path = str(file_path.relative_to(repo_root))
                if not any(mod["path"] == relative_path for mod in integration_map["core_modules"]):
                    analysis = analyze_integration_file(file_path)
                    if analysis.get("exists") and (
                        analysis.get("ollama_methods") or 
                        analysis.get("api_calls") or
                        "ollama" in file_path.read_text(encoding="utf-8", errors="ignore").lower()
                    ):
                        integration_map["core_modules"].append(analysis)
    
    # Build dependency graph
    print("  🕸️ Building dependency graph...")
    integration_map["dependency_graph"] = build_dependency_graph()
    
    # Analyze data flows
    print("  🔄 Analyzing data flows...")
    integration_map["data_flows"] = analyze_data_flows()
    
    # Analyze integration patterns
    print("  🎭 Analyzing integration patterns...")
    pattern_summary = defaultdict(int)
    for module in integration_map["core_modules"]:
        for pattern in module.get("integration_patterns", []):
            pattern_summary[pattern["pattern"]] += 1
    integration_map["integration_patterns"] = dict(pattern_summary)
    
    # Store results
    ollama_analysis["integrations"] = integration_map
    
    return integration_map

# Execute integration mapping
integration_results = map_ollama_integrations()

# Display results
print("\n🔄 OLLAMA INTEGRATION MAPPING RESULTS")
print("=" * 50)
print(f"🏗️ Core Integration Modules: {len(integration_results['core_modules'])}")
print(f"🕸️ Dependencies Mapped: {len(integration_results['dependency_graph'])}")
print(f"🔄 Data Flow Patterns: {len(integration_results['data_flows'])}")
print(f"🎭 Integration Patterns: {len(integration_results['integration_patterns'])}")

# Show core modules
if integration_results["core_modules"]:
    print("\n🏗️ Core Integration Modules:")
    sorted_modules = sorted(integration_results["core_modules"], 
                           key=lambda x: x.get("size", 0), reverse=True)
    
    for module in sorted_modules:
        path = module["path"]
        size = module.get("size", 0)
        lines = module.get("lines", 0)
        classes = len(module.get("classes", []))
        functions = len(module.get("functions", []))
        ollama_methods = len(module.get("ollama_methods", []))
        api_calls = len(module.get("api_calls", []))
        
        print(f"  📄 {path}")
        print(f"    📏 {lines} lines ({size} bytes)")
        if classes:
            print(f"    🏛️ {classes} classes")
        if functions:
            print(f"    ⚙️ {functions} functions ({ollama_methods} Ollama-specific)")
        if api_calls:
            print(f"    🌐 {api_calls} API calls")

# Show integration patterns
if integration_results["integration_patterns"]:
    print("\n🎭 Integration Pattern Usage:")
    for pattern, count in sorted(integration_results["integration_patterns"].items(), 
                                key=lambda x: x[1], reverse=True):
        print(f"  🔧 {pattern}: {count} occurrences")

# Show data flow analysis
if integration_results["data_flows"]:
    print(f"\n🔄 Data Flow Analysis ({len(integration_results['data_flows'])} flows):")
    flow_types = defaultdict(int)
    for flow in integration_results["data_flows"]:
        flow_types[flow["flow_type"]] += 1
    
    for flow_type, count in sorted(flow_types.items(), key=lambda x: x[1], reverse=True):
        print(f"  📊 {flow_type}: {count} flows")

# Show dependency graph summary
if integration_results["dependency_graph"]:
    print("\n🕸️ Dependency Graph Summary:")
    total_deps = sum(len(deps) for deps in integration_results["dependency_graph"].values())
    print(f"  📦 Total Dependencies: {total_deps}")
    print(f"  📁 Files with Dependencies: {len(integration_results['dependency_graph'])}")
    
    # Show top dependencies
    all_deps = []
    for file_deps in integration_results["dependency_graph"].values():
        all_deps.extend(file_deps)
    
    if all_deps:
        dep_counts = defaultdict(int)
        for dep in all_deps:
            dep_counts[dep] += 1
        
        print("  📈 Most Common Dependencies:")
        for dep, count in sorted(dep_counts.items(), key=lambda x: x[1], reverse=True)[:5]:
            print(f"    📋 {dep}: {count} imports")
````

````python
def generate_system_documentation():
    """Generate comprehensive system documentation and knowledge graph"""
    print("📚 Generating comprehensive Ollama system documentation...")
    
    documentation = {
        "system_overview": {},
        "architecture_diagram": {},
        "integration_flows": [],
        "api_documentation": {},
        "configuration_guide": {},
        "troubleshooting": {},
        "obsidian_notes": [],
        "knowledge_graph": {},
        "enhancement_recommendations": []
    }
    
    # Generate system overview
    def create_system_overview():
        """Create comprehensive system overview"""
        overview = {
            "title": "KILO-FOOLISH Ollama Integration System",
            "description": "Quantum-inspired AI orchestration with local Ollama models",
            "components": [],
            "key_features": [],
            "integration_points": [],
            "architecture_layers": []
        }
        
        # Analyze discovered components
        ollama_analysis.get("processes", {})
        file_refs = ollama_analysis.get("file_references", {})
        config_info = ollama_analysis.get("configuration", {})
        integration_info = ollama_analysis.get("integrations", {})
        
        # Key components
        overview["components"] = [
            {
                "name": "Enhanced Ollama Hub",
                "description": "Central Ollama integration with model specialization",
                "file": "src/integration/ollama_integration.py",
                "type": "Core Integration"
            },
            {
                "name": "ChatDev LLM Adapter", 
                "description": "Multi-agent collaborative development with Ollama",
                "file": "src/ai/ollama_chatdev_integrator.py",
                "type": "AI Orchestration"
            },
            {
                "name": "Ollama Model Manager",
                "description": "Model lifecycle and configuration management", 
                "file": "src/ai/ollama_model_manager.py",
                "type": "Model Management"
            },
            {
                "name": "Local AI Coordinator",
                "description": "Privacy-first local AI coordination",
                "file": "src/core/ai_coordinator.py",
                "type": "Core System"
            }
        ]
        
        # Key features from analysis
        overview["key_features"] = [
            "🤖 Local AI Model Integration (Ollama)",
            "🧠 Multi-Agent Collaborative Development (ChatDev)",
            "🔒 Privacy-First Local Processing",
            "⚡ Async/Streaming Response Handling", 
            "🎭 Model Specialization & Role Assignment",
            "🔄 Consciousness Bridge Integration",
            "📊 Real-time Process Monitoring",
            "🏗️ Quantum-Inspired Architecture"
        ]
        
        # Integration points
        integration_count = len(integration_info.get("core_modules", []))
        config_count = len(config_info.get("configuration_files", []))
        reference_count = file_refs.get("total_files", 0)
        
        overview["integration_points"] = [
            f"📄 {integration_count} Core Integration Modules",
            f"⚙️ {config_count} Configuration Files", 
            f"🔍 {reference_count} Files with Ollama References",
            f"🌐 {len(config_info.get('api_endpoints', []))} API Endpoints",
            f"🤖 {len(config_info.get('model_configurations', []))} Model Configurations"
        ]
        
        # Architecture layers
        overview["architecture_layers"] = [
            {
                "layer": "🎮 User Interface Layer",
                "components": ["Jupyter Notebooks", "Command Line", "VS Code Integration"],
                "description": "Interactive development and monitoring interfaces"
            },
            {
                "layer": "🧠 AI Orchestration Layer", 
                "components": ["AI Coordinator", "ChatDev Adapter", "Multi-LLM Orchestra"],
                "description": "AI agent coordination and task routing"
            },
            {
                "layer": "🤖 Model Integration Layer",
                "components": ["Ollama Hub", "Model Manager", "API Adapters"],
                "description": "Local AI model integration and management"
            },
            {
                "layer": "🔄 Consciousness Layer",
                "components": ["Memory Systems", "Context Bridges", "Progress Tracking"],
                "description": "System consciousness and context retention"
            },
            {
                "layer": "💾 Data & Configuration Layer",
                "components": ["Configuration Files", "Model Data", "Logs"],
                "description": "Data persistence and system configuration"
            }
        ]
        
        return overview
    
    def create_obsidian_knowledge_graph():
        """Create Obsidian-compatible knowledge graph notes"""
        obsidian_notes = []
        
        # Main system note
        main_note = {
            "title": "Ollama Integration System",
            "tags": ["#ollama", "#ai", "#integration", "#kilo-foolish"],
            "content": f"""# Ollama Integration System
            
## Overview
The KILO-FOOLISH Ollama Integration System provides quantum-inspired AI orchestration with local model processing.

## Key Components
{chr(10).join(f"- [[{comp['name']}]] - {comp['description']}" for comp in documentation['system_overview']['components'])}

## Architecture Layers
{chr(10).join(f"### {layer['layer']}{chr(10)}{layer['description']}{chr(10)}" for layer in documentation['system_overview']['architecture_layers'])}

## Integration Statistics  
- **Files with Ollama References**: {ollama_analysis.get('file_references', {}).get('total_files', 0)}
- **Configuration Files**: {len(ollama_analysis.get('configuration', {}).get('configuration_files', []))}
- **API Endpoints**: {len(ollama_analysis.get('configuration', {}).get('api_endpoints', []))}
- **Integration Modules**: {len(ollama_analysis.get('integrations', {}).get('core_modules', []))}

## Related Notes
- [[ChatDev Integration]]
- [[AI Coordination]]
- [[Model Management]]
- [[Configuration Guide]]
            """,
            "links": []
        }
        obsidian_notes.append(main_note)
        
        # Component-specific notes
        for component in documentation["system_overview"]["components"]:
            comp_note = {
                "title": component["name"],
                "tags": ["#component", "#ollama", component["type"].lower().replace(" ", "-")],
                "content": f"""# {component['name']}

## Description
{component['description']}

## File Location
`{component['file']}`

## Type
{component['type']}

## Integration Points
- Part of [[Ollama Integration System]]
- Connects to [[AI Coordination]]

## Technical Details
*Detailed technical analysis would be extracted from file content analysis*
                """,
                "links": ["Ollama Integration System", "AI Coordination"]
            }
            obsidian_notes.append(comp_note)
        
        return obsidian_notes
    
    def create_api_documentation():
        """Create API endpoint documentation"""
        api_docs = {
            "base_url": "http://localhost:11434",
            "endpoints": [],
            "authentication": "None (local service)",
            "response_format": "JSON",
            "streaming_support": True
        }
        
        # Standard Ollama API endpoints
        standard_endpoints = [
            {
                "endpoint": "/api/generate",
                "method": "POST", 
                "description": "Generate text completion from a model",
                "parameters": {
                    "model": "string (required) - Name of the model to use",
                    "prompt": "string (required) - Text prompt for generation",
                    "stream": "boolean (optional) - Enable streaming responses",
                    "context": "array (optional) - Conversation context"
                },
                "example": {
                    "request": {
                        "model": "llama3.2",
                        "prompt": "Hello, how are you?",
                        "stream": False
                    }
                }
            },
            {
                "endpoint": "/api/chat",
                "method": "POST",
                "description": "Chat with a model using conversation format",
                "parameters": {
                    "model": "string (required) - Name of the model to use",
                    "messages": "array (required) - Array of message objects",
                    "stream": "boolean (optional) - Enable streaming responses"
                }
            },
            {
                "endpoint": "/api/tags",
                "method": "GET",
                "description": "List available models",
                "parameters": {},
                "response": {
                    "models": "array - List of available models with metadata"
                }
            },
            {
                "endpoint": "/api/pull",
                "method": "POST", 
                "description": "Download a model from the registry",
                "parameters": {
                    "name": "string (required) - Name of the model to pull"
                }
            }
        ]
        
        api_docs["endpoints"] = standard_endpoints
        
        # Add discovered endpoint usage from configuration analysis
        config_endpoints = ollama_analysis.get("configuration", {}).get("api_endpoints", [])
        if config_endpoints:
            api_docs["discovered_usage"] = []
            endpoint_usage = defaultdict(int)
            for endpoint in config_endpoints:
                endpoint_usage[endpoint["endpoint"]] += 1
            
            for endpoint, count in endpoint_usage.items():
                api_docs["discovered_usage"].append({
                    "endpoint": endpoint,
                    "usage_count": count,
                    "description": f"Found {count} references in codebase"
                })
        
        return api_docs
    
    def create_troubleshooting_guide():
        """Create troubleshooting guide based on analysis"""
        guide = {
            "common_issues": [],
            "connectivity_problems": [],
            "model_issues": [],
            "configuration_problems": [],
            "performance_optimization": []
        }
        
        # Common issues based on analysis
        conn_status = ollama_analysis.get("configuration", {}).get("connection_details", {}).get("status")
        
        if conn_status == "disconnected":
            guide["connectivity_problems"].append({
                "issue": "Ollama service not running",
                "symptoms": ["Connection refused errors", "API endpoints not accessible"],
                "solutions": [
                    "Start Ollama service: `ollama serve`",
                    "Check if port 11434 is available",
                    "Verify Ollama installation"
                ]
            })
        
        guide["common_issues"] = [
            {
                "issue": "Import errors for Ollama modules",
                "symptoms": ["ModuleNotFoundError: No module named 'ollama'"],
                "solutions": [
                    "Install Ollama Python client: `pip install ollama`",
                    "Check Python environment configuration",
                    "Verify virtual environment activation"
                ]
            },
            {
                "issue": "Model not found errors",
                "symptoms": ["Model 'model_name' not found"],
                "solutions": [
                    "Pull model: `ollama pull model_name`",
                    "List available models: `ollama list`",
                    "Check model name spelling"
                ]
            }
        ]
        
        return guide
    
    def generate_enhancement_recommendations():
        """Generate recommendations for system enhancement"""
        recommendations = []
        
        # Analyze current system state
        len(ollama_analysis.get("integrations", {}).get("core_modules", []))
        config_files = len(ollama_analysis.get("configuration", {}).get("configuration_files", []))
        
        # Recommendations based on analysis
        recommendations = [
            {
                "category": "🔄 Integration Enhancement",
                "priority": "High",
                "recommendation": "Implement unified Ollama configuration management",
                "rationale": f"Found {config_files} configuration files - consolidation would improve maintainability",
                "implementation": [
                    "Create centralized config schema",
                    "Implement config validation",
                    "Add hot-reloading capability"
                ]
            },
            {
                "category": "📊 Monitoring & Observability", 
                "priority": "Medium",
                "recommendation": "Add comprehensive Ollama metrics and monitoring",
                "rationale": "Enhanced observability for performance optimization and debugging",
                "implementation": [
                    "Add response time metrics",
                    "Implement model usage tracking",
                    "Create performance dashboards"
                ]
            },
            {
                "category": "🧠 AI Enhancement",
                "priority": "High", 
                "recommendation": "Implement advanced context management",
                "rationale": "Leverage KILO-FOOLISH consciousness system for better AI coordination",
                "implementation": [
                    "Integrate with consciousness bridge",
                    "Add persistent context storage",
                    "Implement context compression algorithms"
                ]
            },
            {
                "category": "🔒 Security & Privacy",
                "priority": "Medium",
                "recommendation": "Add security layers for model access",
                "rationale": "Enhance security for production deployments",
                "implementation": [
                    "Implement API key management",
                    "Add request rate limiting",
                    "Create audit logging"
                ]
            }
        ]
        
        return recommendations
    
    # Generate all documentation components
    print("  📋 Creating system overview...")
    documentation["system_overview"] = create_system_overview()
    
    print("  🕸️ Creating knowledge graph...")
    documentation["obsidian_notes"] = create_obsidian_knowledge_graph()
    
    print("  📖 Creating API documentation...")
    documentation["api_documentation"] = create_api_documentation()
    
    print("  🔧 Creating troubleshooting guide...")
    documentation["troubleshooting"] = create_troubleshooting_guide()
    
    print("  💡 Generating enhancement recommendations...")
    documentation["enhancement_recommendations"] = generate_enhancement_recommendations()
    
    # Store complete documentation
    ollama_analysis["documentation"] = documentation
    
    return documentation

# Generate comprehensive documentation
doc_results = generate_system_documentation()

# Display documentation summary
print("\n📚 SYSTEM DOCUMENTATION GENERATED")
print("=" * 50)
print("📋 System Overview: ✅ Complete")
print(f"🕸️ Obsidian Notes: {len(doc_results['obsidian_notes'])} notes")
print(f"📖 API Endpoints: {len(doc_results['api_documentation']['endpoints'])} documented")
print(f"🔧 Troubleshooting: {len(doc_results['troubleshooting']['common_issues'])} issues covered")
print(f"💡 Recommendations: {len(doc_results['enhancement_recommendations'])} enhancements")

# Show system overview
overview = doc_results["system_overview"]
print(f"\n📋 System Overview: {overview['title']}")
print(f"   {overview['description']}")
print("\n🏗️ Architecture Components:")
for component in overview["components"]:
    print(f"  🔧 {component['name']} ({component['type']})")
    print(f"     📄 {component['file']}")

# Show key features
print("\n✨ Key Features:")
for feature in overview["key_features"]:
    print(f"  {feature}")

# Show enhancement recommendations
print("\n💡 Top Enhancement Recommendations:")
for rec in doc_results["enhancement_recommendations"][:3]:  # Top 3
    priority_emoji = "🔴" if rec["priority"] == "High" else "🟡" if rec["priority"] == "Medium" else "🟢"
    print(f"  {priority_emoji} {rec['category']}: {rec['recommendation']}")
    print(f"     💭 {rec['rationale']}")

# Show Obsidian integration
print("\n🕸️ Obsidian Knowledge Graph:")
print(f"  📝 Generated {len(doc_results['obsidian_notes'])} interconnected notes")
print("  🏷️ Tags: #ollama, #ai, #integration, #kilo-foolish")
print("  🔗 Cross-referenced with system components and architecture")

print("\n✅ Comprehensive Ollama system mapping complete!")
print("📊 Analysis stored in ollama_analysis dictionary for further processing")
````

````python
# Interactive Analysis Functions
def explore_system_component(component_name: str):
    """Explore specific system component in detail"""
    print(f"🔍 Exploring: {component_name}")
    
    # Search in integration modules
    integrations = ollama_analysis.get("integrations", {}).get("core_modules", [])
    for module in integrations:
        if component_name.lower() in module.get("path", "").lower():
            print(f"\n📄 Found in: {module['path']}")
            print(f"📏 Size: {module.get('lines', 0)} lines ({module.get('size', 0)} bytes)")
            
            classes = module.get("classes", [])
            if classes:
                print(f"🏛️ Classes ({len(classes)}):")
                for cls in classes[:5]:  # Top 5
                    print(f"  - {cls['name']} (line {cls['line']})")
            
            functions = module.get("functions", [])
            if functions:
                print(f"⚙️ Functions ({len(functions)}):")
                for func in functions[:5]:  # Top 5
                    print(f"  - {func['name']} (line {func['line']})")
            
            api_calls = module.get("api_calls", [])
            if api_calls:
                print(f"🌐 API Calls ({len(api_calls)}):")
                for api in api_calls[:3]:  # Top 3
                    print(f"  - Line {api['line']}: {api['content'][:50]}...")
            
            break
    else:
        print(f"❌ Component '{component_name}' not found in integration modules")

def query_analysis_data(query: str):
    """Query the analysis data with natural language"""
    query_lower = query.lower()
    results = []
    
    # Search across all analysis data
    if "files" in query_lower or "count" in query_lower:
        file_refs = ollama_analysis.get("file_references", {})
        results.append(f"📁 Files with Ollama references: {file_refs.get('total_files', 0)}")
        results.append(f"🔍 Total references found: {file_refs.get('total_references', 0)}")
    
    if "config" in query_lower or "setting" in query_lower:
        config = ollama_analysis.get("configuration", {})
        results.append(f"⚙️ Configuration files: {len(config.get('configuration_files', []))}")
        results.append(f"🔗 API endpoints: {len(config.get('api_endpoints', []))}")
    
    if "process" in query_lower or "running" in query_lower:
        processes = ollama_analysis.get("processes", {})
        running = len([p for p in processes.get("all_processes", []) if "ollama" in p.get("name", "").lower()])
        results.append(f"🏃 Ollama processes running: {running}")
    
    if "integration" in query_lower or "module" in query_lower:
        integrations = ollama_analysis.get("integrations", {})
        results.append(f"🏗️ Integration modules: {len(integrations.get('core_modules', []))}")
        results.append(f"🕸️ Dependencies mapped: {len(integrations.get('dependency_graph', {}))}")
    
    if results:
        print(f"🔍 Query: '{query}'")
        for result in results:
            print(f"  {result}")
    else:
        print(f"❓ No specific results for query: '{query}'")
        print("💡 Try queries like: 'files count', 'config settings', 'running processes', 'integration modules'")

def generate_system_report():
    """Generate final comprehensive system report"""
    print("📊 COMPREHENSIVE OLLAMA SYSTEM REPORT")
    print("=" * 60)
    
    # System Status
    print("🚦 SYSTEM STATUS")
    processes = ollama_analysis.get("processes", {})
    config = ollama_analysis.get("configuration", {})
    
    ollama_running = any("ollama" in p.get("name", "").lower() for p in processes.get("all_processes", []))
    connection_status = config.get("connection_details", {}).get("status", "unknown")
    
    print(f"  🏃 Ollama Process: {'✅ Running' if ollama_running else '❌ Not detected'}")
    print(f"  🔗 API Connection: {'✅ Connected' if connection_status == 'connected' else '❌ ' + connection_status.title()}")
    
    models = config.get("connection_details", {}).get("models", [])
    if models:
        print(f"  🤖 Available Models: {len(models)} ({', '.join(models[:3])}{'...' if len(models) > 3 else ''})")
    
    # Architecture Overview
    print("\n🏗️ ARCHITECTURE OVERVIEW")
    file_refs = ollama_analysis.get("file_references", {})
    integrations = ollama_analysis.get("integrations", {})
    
    print(f"  📁 Total Files with References: {file_refs.get('total_files', 0)}")
    print(f"  🐍 Python Integration Files: {len(file_refs.get('python_files', []))}")
    print(f"  ⚙️ Configuration Files: {len(config.get('configuration_files', []))}")
    print(f"  🏗️ Core Integration Modules: {len(integrations.get('core_modules', []))}")
    
    # Key Integration Points
    print("\n🔗 KEY INTEGRATION POINTS")
    core_modules = integrations.get("core_modules", [])
    if core_modules:
        sorted_modules = sorted(core_modules, key=lambda x: x.get("size", 0), reverse=True)[:5]
        for i, module in enumerate(sorted_modules, 1):
            size_kb = module.get("size", 0) / 1024
            functions = len(module.get("functions", []))
            api_calls = len(module.get("api_calls", []))
            print(f"  {i}. {module['path']} ({size_kb:.1f}KB, {functions} functions, {api_calls} API calls)")
    
    # Configuration Analysis
    print("\n⚙️ CONFIGURATION ANALYSIS")
    api_endpoints = config.get("api_endpoints", [])
    if api_endpoints:
        endpoint_types = defaultdict(int)
        for endpoint in api_endpoints:
            endpoint_types[endpoint.get("endpoint", "unknown")] += 1
        
        print("  🌐 API Endpoint Usage:")
        for endpoint, count in sorted(endpoint_types.items(), key=lambda x: x[1], reverse=True)[:5]:
            print(f"    - {endpoint}: {count} references")
    
    # Integration Patterns
    print("\n🎭 INTEGRATION PATTERNS")
    patterns = integrations.get("integration_patterns", {})
    if patterns:
        for pattern, count in sorted(patterns.items(), key=lambda x: x[1], reverse=True)[:3]:
            print(f"  🔧 {pattern}: {count} implementations")
    
    # System Health & Recommendations
    print("\n💊 SYSTEM HEALTH & RECOMMENDATIONS")
    doc = ollama_analysis.get("documentation", {})
    recommendations = doc.get("enhancement_recommendations", [])
    
    high_priority = [r for r in recommendations if r.get("priority") == "High"]
    print(f"  🔴 High Priority Enhancements: {len(high_priority)}")
    for rec in high_priority[:2]:
        print(f"    - {rec['recommendation']}")
    
    medium_priority = [r for r in recommendations if r.get("priority") == "Medium"]
    print(f"  🟡 Medium Priority Enhancements: {len(medium_priority)}")
    
    # Knowledge Graph Summary
    print("\n🕸️ KNOWLEDGE GRAPH INTEGRATION")
    obsidian_notes = doc.get("obsidian_notes", [])
    print(f"  📝 Obsidian Notes Generated: {len(obsidian_notes)}")
    print("  🏷️ Tagged with: #ollama, #ai, #integration, #kilo-foolish")
    print("  🔗 Cross-referenced architecture components and flows")
    
    print("\n✅ ANALYSIS COMPLETE - System mapped and documented!")
    return ollama_analysis

# Execute interactive analysis
print("🎯 INTERACTIVE OLLAMA SYSTEM EXPLORATION")
print("=" * 50)

# Generate final comprehensive report
final_report = generate_system_report()

print("\n🔍 Interactive Functions Available:")
print("  📋 explore_system_component('component_name') - Deep dive into specific components")
print("  🔍 query_analysis_data('your query') - Natural language queries of analysis data")
print("  📊 generate_system_report() - Regenerate comprehensive report")
print("  📚 ollama_analysis - Complete analysis data dictionary")

print("\n💡 Example Explorations:")
print("  explore_system_component('ollama_integration')")
print("  explore_system_component('chatdev')")
print("  query_analysis_data('how many files use ollama?')")
print("  query_analysis_data('configuration settings found')")

print("\n🎯 Ready for interactive exploration!")
print("📝 Use this notebook to continue mapping, documenting, and developing the system")
print("🔄 All analysis data is preserved in the ollama_analysis dictionary")
````
