## 📋 Documentation Ecosystem Overview

The `docs/` directory serves as the central knowledge repository for the KILO-FOOLISH system, containing guides, reports, notebooks, and integration documentation.

```python
import json
import re
from collections import defaultdict
from datetime import datetime
from pathlib import Path

# Initialize documentation analysis
docs_root = Path("../../docs")
analysis_data = {
    "timestamp": datetime.now().isoformat(),
    "directory": "docs/",
    "subdirectories": [],
    "documentation_files": [],
    "file_types": defaultdict(int),
    "documentation_categories": {},
    "knowledge_graph": {},
    "notebook_analysis": {}
}

print("📚 Analyzing docs/ directory structure...")

# Scan directory structure recursively
def scan_docs_recursive(path, relative_to=None):
    if relative_to is None:
        relative_to = docs_root.parent
    
    items = {"files": [], "subdirs": []}
    
    if path.exists():
        for item in path.iterdir():
            if item.is_dir():
                subdir_info = {
                    "name": item.name,
                    "path": str(item.relative_to(relative_to)),
                    "files_count": len(list(item.rglob("*"))),
                    "content": scan_docs_recursive(item, relative_to)
                }
                items["subdirs"].append(subdir_info)
                analysis_data["subdirectories"].append(subdir_info)
            else:
                file_info = {
                    "name": item.name,
                    "path": str(item.relative_to(relative_to)),
                    "size": item.stat().st_size,
                    "type": item.suffix,
                    "category": categorize_doc_file(item)
                }
                items["files"].append(file_info)
                analysis_data["documentation_files"].append(file_info)
                analysis_data["file_types"][item.suffix] += 1
    
    return items

def categorize_doc_file(file_path):
    """Categorize documentation files based on name and extension"""
    name_lower = file_path.name.lower()
    
    if file_path.suffix == ".ipynb":
        return "jupyter_notebook"
    elif file_path.suffix == ".md":
        if "readme" in name_lower:
            return "readme"
        elif "guide" in name_lower or "tutorial" in name_lower:
            return "guide"
        elif "report" in name_lower or "analysis" in name_lower:
            return "report"
        elif "spec" in name_lower or "specification" in name_lower:
            return "specification"
        else:
            return "documentation"
    elif file_path.suffix in [".txt", ".log"]:
        return "log_or_text"
    elif file_path.suffix in [".json", ".yaml", ".yml"]:
        return "structured_data"
    elif file_path.suffix in [".png", ".jpg", ".jpeg", ".gif", ".svg"]:
        return "image"
    else:
        return "other"

# Perform recursive scan
docs_structure = scan_docs_recursive(docs_root)

print(f"📁 Found {len(analysis_data['subdirectories'])} subdirectories")
print(f"📄 Found {len(analysis_data['documentation_files'])} documentation files")

# Display structure summary
for subdir in analysis_data["subdirectories"]:
    print(f"  📂 {subdir['name']}: {subdir['files_count']} items")
```

## 📊 Documentation Analysis by Category

```python
# Analyze documentation by category
doc_categories = defaultdict(list)
for file_info in analysis_data["documentation_files"]:
    doc_categories[file_info["category"]].append(file_info)

print("📊 DOCUMENTATION ANALYSIS BY CATEGORY")
print("=" * 40)

category_descriptions = {
    "jupyter_notebook": "📓 Interactive analysis and development notebooks",
    "readme": "📖 Project and module introduction documents",
    "guide": "📚 Step-by-step guides and tutorials",
    "report": "📋 Analysis reports and system assessments",
    "specification": "📐 Technical specifications and standards",
    "documentation": "📝 General documentation and references",
    "log_or_text": "📄 Log files and plain text documents",
    "structured_data": "🗃️ JSON/YAML data files",
    "image": "🖼️ Images, diagrams, and visual content",
    "other": "📎 Other file types"
}

total_size = 0
for category, files in doc_categories.items():
    category_size = sum(f["size"] for f in files)
    total_size += category_size
    
    print(f"\n{category_descriptions.get(category, f'📁 {category.title()}')}")
    print(f"📊 {len(files)} files, {category_size / 1024:.1f} KB")
    
    # Show largest files in category
    if files:
        sorted_files = sorted(files, key=lambda x: x["size"], reverse=True)[:3]
        for file_info in sorted_files:
            size_kb = file_info["size"] / 1024
            print(f"  📄 {file_info['name']} ({size_kb:.1f} KB)")

print(f"\n📊 Total Documentation Size: {total_size / 1024:.1f} KB")
analysis_data["documentation_categories"] = dict(doc_categories)
analysis_data["total_size_kb"] = total_size / 1024
```

## 📓 Jupyter Notebook Analysis

```python
# Analyze Jupyter notebooks specifically
notebooks = doc_categories.get("jupyter_notebook", [])
notebook_analysis = {
    "total_notebooks": len(notebooks),
    "notebook_details": [],
    "notebook_categories": defaultdict(list),
    "total_notebook_size": sum(nb["size"] for nb in notebooks)
}

print("📓 JUPYTER NOTEBOOK ANALYSIS")
print("=" * 40)

if notebooks:
    for notebook in notebooks:
        # Try to analyze notebook content
        notebook_path = docs_root.parent / notebook["path"]
        notebook_info = {
            "name": notebook["name"],
            "path": notebook["path"],
            "size_kb": notebook["size"] / 1024,
            "cells": 0,
            "category": "unknown"
        }
        
        try:
            if notebook_path.exists():
                with open(notebook_path, "r", encoding="utf-8") as f:
                    notebook_content = json.load(f)
                
                notebook_info["cells"] = len(notebook_content.get("cells", []))
                
                # Categorize notebook based on name/content
                name_lower = notebook["name"].lower()
                if "directory" in name_lower:
                    notebook_info["category"] = "directory_analysis"
                elif "ollama" in name_lower or "ai" in name_lower:
                    notebook_info["category"] = "ai_integration"
                elif "system" in name_lower or "analysis" in name_lower:
                    notebook_info["category"] = "system_analysis"
                elif "quantum" in name_lower:
                    notebook_info["category"] = "quantum_development"
                else:
                    notebook_info["category"] = "general"
                
        except Exception as e:
            notebook_info["error"] = str(e)
        
        notebook_analysis["notebook_details"].append(notebook_info)
        notebook_analysis["notebook_categories"][notebook_info["category"]].append(notebook_info)
        
        print(f"📓 {notebook_info['name']}")
        print(f"   📊 {notebook_info['cells']} cells, {notebook_info['size_kb']:.1f} KB")
        print(f"   🏷️ Category: {notebook_info['category']}")
        if "error" in notebook_info:
            print(f"   ⚠️ Analysis error: {notebook_info['error']}")
        print()
    
    # Summary by category
    print("📊 NOTEBOOK CATEGORIES SUMMARY")
    print("=" * 30)
    for category, category_notebooks in notebook_analysis["notebook_categories"].items():
        total_cells = sum(nb["cells"] for nb in category_notebooks)
        total_size = sum(nb["size_kb"] for nb in category_notebooks)
        print(f"🏷️ {category.replace('_', ' ').title()}: {len(category_notebooks)} notebooks")
        print(f"   📊 {total_cells} total cells, {total_size:.1f} KB")

else:
    print("📓 No Jupyter notebooks found in docs directory")

analysis_data["notebook_analysis"] = notebook_analysis
```

## 🕸️ Knowledge Graph Analysis

```python
# Analyze knowledge connections and cross-references
knowledge_graph = {
    "cross_references": [],
    "topic_clusters": defaultdict(list),
    "interconnections": 0,
    "orphaned_docs": [],
    "hub_documents": []
}

print("🕸️ KNOWLEDGE GRAPH ANALYSIS")
print("=" * 40)

# Define topic keywords for clustering
topic_keywords = {
    "quantum": ["quantum", "consciousness", "bridge", "entanglement"],
    "ai_integration": ["ollama", "chatdev", "ai", "llm", "model"],
    "system_management": ["process", "system", "management", "coordination"],
    "development": ["development", "coding", "programming", "workflow"],
    "configuration": ["config", "settings", "environment", "setup"],
    "analysis": ["analysis", "report", "assessment", "audit"]
}

# Analyze each document for topics and connections
for file_info in analysis_data["documentation_files"]:
    if file_info["type"] in [".md", ".txt"]:
        file_path = docs_root.parent / file_info["path"]
        
        try:
            if file_path.exists() and file_path.stat().st_size < 1024*1024:  # Only analyze files < 1MB
                with open(file_path, "r", encoding="utf-8", errors="ignore") as f:
                    content = f.read().lower()
                
                # Find topic matches
                file_topics = []
                for topic, keywords in topic_keywords.items():
                    if any(keyword in content for keyword in keywords):
                        file_topics.append(topic)
                        knowledge_graph["topic_clusters"][topic].append(file_info["name"])
                
                # Look for cross-references (markdown links, file references)
                md_links = re.findall(r"\[.*?\]\(.*?\.md\)|\[.*?\]\(.*?\.ipynb\)", content)
                if md_links:
                    knowledge_graph["cross_references"].extend([
                        {"source": file_info["name"], "target": link, "type": "markdown_link"}
                        for link in md_links
                    ])
                    knowledge_graph["interconnections"] += len(md_links)
                
                # Check if it's a hub document (mentions many other files)
                file_mentions = len(re.findall(r"\b\w+\.(md|ipynb|py|json)\b", content))
                if file_mentions > 5:
                    knowledge_graph["hub_documents"].append({
                        "name": file_info["name"],
                        "mentions": file_mentions,
                        "topics": file_topics
                    })
                
                # Check for orphaned documents (no links to/from other docs)
                if not md_links and file_mentions < 2:
                    knowledge_graph["orphaned_docs"].append(file_info["name"])
                
        except Exception as e:
            print(f"⚠️ Could not analyze {file_info['name']}: {e}")

# Display knowledge graph analysis
print(f"🔗 Cross-references found: {knowledge_graph['interconnections']}")
print(f"📚 Hub documents: {len(knowledge_graph['hub_documents'])}")
print(f"🏝️ Orphaned documents: {len(knowledge_graph['orphaned_docs'])}")

print("\n📊 TOPIC CLUSTERS")
print("=" * 20)
for topic, files in knowledge_graph["topic_clusters"].items():
    print(f"🏷️ {topic.replace('_', ' ').title()}: {len(files)} documents")
    if len(files) <= 5:
        for file in files:
            print(f"   📄 {file}")
    else:
        for file in files[:3]:
            print(f"   📄 {file}")
        print(f"   📄 ... and {len(files)-3} more")
    print()

if knowledge_graph["hub_documents"]:
    print("🌟 HUB DOCUMENTS (High Interconnectivity)")
    print("=" * 40)
    for hub in sorted(knowledge_graph["hub_documents"], key=lambda x: x["mentions"], reverse=True):
        print(f"📚 {hub['name']} - {hub['mentions']} file mentions")
        print(f"   🏷️ Topics: {', '.join(hub['topics'])}")

analysis_data["knowledge_graph"] = knowledge_graph
```

## 🎯 Documentation Recommendations

```python
# Generate documentation improvement recommendations
doc_recommendations = []

# Check for orphaned documents
if knowledge_graph["orphaned_docs"]:
    doc_recommendations.append({
        "priority": "🟡 Medium",
        "category": "Knowledge Graph Integration",
        "issue": f"{len(knowledge_graph['orphaned_docs'])} documents appear orphaned",
        "recommendation": "Create cross-references and integrate into knowledge graph",
        "implementation": "Add links to/from related documents, create index pages",
        "affected_files": knowledge_graph["orphaned_docs"][:5]
    })

# Check notebook coverage
if notebook_analysis["total_notebooks"] < len(analysis_data["subdirectories"]):
    missing_notebooks = len(analysis_data["subdirectories"]) - notebook_analysis["total_notebooks"]
    doc_recommendations.append({
        "priority": "🔵 Enhancement",
        "category": "Notebook Coverage",
        "issue": f"Only {notebook_analysis['total_notebooks']} notebooks for {len(analysis_data['subdirectories'])} directories",
        "recommendation": "Create directory-specific analysis notebooks",
        "implementation": f"Generate {missing_notebooks} additional notebooks for comprehensive coverage"
    })

# Check for documentation balance
guides_count = len(doc_categories.get("guide", []))
reports_count = len(doc_categories.get("report", []))
if reports_count > guides_count * 2:
    doc_recommendations.append({
        "priority": "🟢 Low",
        "category": "Documentation Balance",
        "issue": f"{reports_count} reports vs {guides_count} guides - report-heavy",
        "recommendation": "Create more user-friendly guides and tutorials",
        "implementation": "Convert technical reports into step-by-step guides"
    })

# Check for Obsidian integration opportunities
md_files = len(doc_categories.get("documentation", [])) + len(doc_categories.get("guide", [])) + len(doc_categories.get("readme", []))
if md_files > 10:
    doc_recommendations.append({
        "priority": "🔵 Enhancement",
        "category": "Obsidian Integration",
        "issue": f"{md_files} markdown files - good candidate for Obsidian vault",
        "recommendation": "Enhance cross-linking for Obsidian knowledge graph",
        "implementation": "Add [[wikilinks]], tags, and MOCs (Maps of Content)"
    })

# Quantum-specific recommendations
quantum_docs = knowledge_graph["topic_clusters"].get("quantum", [])
if quantum_docs:
    doc_recommendations.append({
        "priority": "🔵 Quantum-Specific",
        "category": "Quantum Documentation",
        "issue": f"{len(quantum_docs)} quantum-related documents detected",
        "recommendation": "Create quantum development guide and consciousness bridge documentation",
        "implementation": "Consolidate quantum concepts into comprehensive guide"
    })

print("🎯 DOCUMENTATION RECOMMENDATIONS")
print("=" * 50)

for i, rec in enumerate(doc_recommendations, 1):
    print(f"{i}. {rec['priority']} - {rec['category']}")
    print(f"   📋 Issue: {rec['issue']}")
    print(f"   💡 Recommendation: {rec['recommendation']}")
    print(f"   🔧 Implementation: {rec['implementation']}")
    if "affected_files" in rec and rec["affected_files"]:
        print(f"   📁 Sample Files: {', '.join(rec['affected_files'])}")
    print()

analysis_data["doc_recommendations"] = doc_recommendations

# Generate comprehensive statistics
doc_stats = {
    "total_files": len(analysis_data["documentation_files"]),
    "total_subdirs": len(analysis_data["subdirectories"]),
    "total_size_mb": analysis_data["total_size_kb"] / 1024,
    "notebooks": notebook_analysis["total_notebooks"],
    "cross_references": knowledge_graph["interconnections"],
    "topic_clusters": len(knowledge_graph["topic_clusters"]),
    "hub_documents": len(knowledge_graph["hub_documents"]),
    "orphaned_docs": len(knowledge_graph["orphaned_docs"])
}

print("📊 DOCUMENTATION STATISTICS SUMMARY")
print("=" * 40)
for key, value in doc_stats.items():
    print(f"📈 {key.replace('_', ' ').title()}: {value}")

analysis_data["doc_stats"] = doc_stats
```
