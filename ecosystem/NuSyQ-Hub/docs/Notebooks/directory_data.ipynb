## 📋 Data Management Overview

The `data/` directory serves as the persistent storage layer for the KILO-FOOLISH system, containing logs, consciousness memory, quest data, and system state information.

```python
import json
import time
from collections import defaultdict
from datetime import datetime
from pathlib import Path

# Initialize data analysis
data_root = Path("../../data")
analysis_data = {
    "timestamp": datetime.now().isoformat(),
    "directory": "data/",
    "subdirectories": [],
    "data_files": [],
    "file_types": defaultdict(int),
    "data_categories": {},
    "storage_analysis": {},
    "consciousness_data": {},
    "log_analysis": {}
}

print("💾 Analyzing data/ directory structure...")

# Scan directory structure recursively
def scan_data_recursive(path, relative_to=None):
    if relative_to is None:
        relative_to = data_root.parent
    
    items = {"files": [], "subdirs": []}
    
    if path.exists():
        for item in path.iterdir():
            if item.is_dir():
                subdir_info = {
                    "name": item.name,
                    "path": str(item.relative_to(relative_to)),
                    "files_count": len(list(item.iterdir())) if item.exists() else 0,
                    "total_size": sum(f.stat().st_size for f in item.rglob("*") if f.is_file()),
                    "purpose": categorize_data_directory(item)
                }
                items["subdirs"].append(subdir_info)
                analysis_data["subdirectories"].append(subdir_info)
                
                # Recursively scan subdirectory
                subdir_content = scan_data_recursive(item, relative_to)
                subdir_info["content"] = subdir_content
                
            else:
                file_info = {
                    "name": item.name,
                    "path": str(item.relative_to(relative_to)),
                    "size": item.stat().st_size,
                    "type": item.suffix,
                    "modified": datetime.fromtimestamp(item.stat().st_mtime).isoformat(),
                    "category": categorize_data_file(item),
                    "age_hours": (time.time() - item.stat().st_mtime) / 3600
                }
                items["files"].append(file_info)
                analysis_data["data_files"].append(file_info)
                analysis_data["file_types"][item.suffix] += 1
    
    return items

def categorize_data_directory(dir_path):
    """Categorize data directories based on name and purpose"""
    name_lower = dir_path.name.lower()
    
    if "log" in name_lower:
        return "system_logging"
    elif "consciousness" in name_lower or "memory" in name_lower:
        return "consciousness_storage"
    elif "quest" in name_lower or "rpg" in name_lower:
        return "quest_system"
    elif "config" in name_lower or "settings" in name_lower:
        return "configuration_data"
    elif "temp" in name_lower or "cache" in name_lower:
        return "temporary_storage"
    elif "backup" in name_lower or "archive" in name_lower:
        return "backup_storage"
    else:
        return "general_storage"

def categorize_data_file(file_path):
    """Categorize data files based on name and extension"""
    name_lower = file_path.name.lower()
    
    if file_path.suffix == ".json":
        if "log" in name_lower:
            return "json_log"
        elif "quest" in name_lower or "rpg" in name_lower:
            return "quest_data"
        elif "config" in name_lower or "settings" in name_lower:
            return "config_data"
        elif "consciousness" in name_lower or "memory" in name_lower:
            return "consciousness_data"
        else:
            return "structured_data"
    elif file_path.suffix in [".jsonl", ".ndjson"]:
        return "streaming_log"
    elif file_path.suffix in [".log", ".txt"]:
        return "text_log"
    elif file_path.suffix in [".db", ".sqlite", ".sqlite3"]:
        return "database"
    elif file_path.suffix in [".yaml", ".yml"]:
        return "yaml_config"
    elif file_path.suffix in [".csv", ".tsv"]:
        return "tabular_data"
    elif file_path.suffix in [".pkl", ".pickle"]:
        return "serialized_data"
    else:
        return "other_data"

# Perform recursive scan
data_structure = scan_data_recursive(data_root)

print(f"📁 Found {len(analysis_data['subdirectories'])} subdirectories")
print(f"📄 Found {len(analysis_data['data_files'])} data files")

# Display structure summary
total_size = 0
for subdir in analysis_data["subdirectories"]:
    size_mb = subdir["total_size"] / (1024 * 1024)
    total_size += subdir["total_size"]
    print(f"  📂 {subdir['name']}: {subdir['files_count']} files, {size_mb:.2f} MB ({subdir['purpose']})")

print(f"\n💾 Total Data Storage: {total_size / (1024 * 1024):.2f} MB")
```

## 📊 Data Analysis by Category

```python
# Analyze data by category
data_categories = defaultdict(list)
for file_info in analysis_data["data_files"]:
    data_categories[file_info["category"]].append(file_info)

print("📊 DATA ANALYSIS BY CATEGORY")
print("=" * 40)

category_descriptions = {
    "json_log": "📋 JSON-formatted log files",
    "streaming_log": "📊 Streaming/append-only log files (.jsonl)",
    "text_log": "📄 Plain text log files",
    "quest_data": "🎮 Quest system and RPG data",
    "consciousness_data": "🧠 Consciousness bridge and memory data",
    "config_data": "⚙️ Configuration and settings data",
    "structured_data": "🗃️ General structured data files",
    "database": "💾 Database files",
    "yaml_config": "📝 YAML configuration files",
    "tabular_data": "📈 CSV/tabular data files",
    "serialized_data": "🔬 Serialized/pickled data",
    "other_data": "📎 Other data file types"
}

storage_analysis = {
    "total_files": len(analysis_data["data_files"]),
    "total_size_mb": sum(f["size"] for f in analysis_data["data_files"]) / (1024 * 1024),
    "categories": {}
}

for category, files in data_categories.items():
    category_size = sum(f["size"] for f in files)
    category_size_mb = category_size / (1024 * 1024)
    
    # Analyze file ages
    recent_files = [f for f in files if f["age_hours"] < 24]  # Less than 24 hours old
    old_files = [f for f in files if f["age_hours"] > 24 * 7]  # Older than 1 week
    
    category_info = {
        "file_count": len(files),
        "total_size_mb": category_size_mb,
        "recent_files": len(recent_files),
        "old_files": len(old_files),
        "avg_file_size_kb": (category_size / len(files)) / 1024 if files else 0
    }
    
    storage_analysis["categories"][category] = category_info
    
    print(f"\n{category_descriptions.get(category, f'📁 {category.title()}')}")
    print(f"📊 {len(files)} files, {category_size_mb:.2f} MB")
    print(f"🕒 {len(recent_files)} recent (< 24h), {len(old_files)} old (> 1 week)")
    print(f"📏 Average file size: {category_info['avg_file_size_kb']:.1f} KB")
    
    # Show largest files in category
    if files:
        sorted_files = sorted(files, key=lambda x: x["size"], reverse=True)[:3]
        for file_info in sorted_files:
            size_kb = file_info["size"] / 1024
            age_desc = "recent" if file_info["age_hours"] < 24 else "old" if file_info["age_hours"] > 168 else "moderate"
            print(f"  📄 {file_info['name']} ({size_kb:.1f} KB, {age_desc})")

analysis_data["storage_analysis"] = storage_analysis
```

## 🧠 Consciousness Data Analysis

```python
# Special analysis for consciousness-related data
consciousness_files = data_categories.get("consciousness_data", [])
quest_files = data_categories.get("quest_data", [])

consciousness_analysis = {
    "consciousness_files": len(consciousness_files),
    "quest_files": len(quest_files),
    "total_consciousness_size": sum(f["size"] for f in consciousness_files + quest_files),
    "memory_patterns": [],
    "quest_progression": [],
    "consciousness_health": "unknown"
}

print("🧠 CONSCIOUSNESS DATA ANALYSIS")
print("=" * 40)

if consciousness_files or quest_files:
    all_consciousness_files = consciousness_files + quest_files
    
    print(f"🧠 Consciousness Files: {len(consciousness_files)}")
    print(f"🎮 Quest Files: {len(quest_files)}")
    print(f"💾 Total Consciousness Storage: {consciousness_analysis['total_consciousness_size'] / 1024:.1f} KB")
    
    # Analyze specific consciousness files
    for file_info in all_consciousness_files:
        file_path = data_root.parent / file_info["path"]
        
        print(f"\n📄 {file_info['name']}")
        print(f"   📊 {file_info['size'] / 1024:.1f} KB, modified {file_info['age_hours']:.1f} hours ago")
        
        # Try to analyze content if it's JSON
        if file_info["type"] == ".json" and file_path.exists():
            try:
                with open(file_path, "r", encoding="utf-8") as f:
                    content = json.load(f)
                
                if isinstance(content, dict):
                    print(f"   🔑 Keys: {', '.join(list(content.keys())[:5])}{'...' if len(content.keys()) > 5 else ''}")
                    
                    # Look for consciousness patterns
                    if "consciousness" in str(content).lower():
                        consciousness_analysis["memory_patterns"].append(file_info["name"])
                        print("   🧠 Contains consciousness data")
                    
                    if "quest" in str(content).lower() or "progress" in str(content).lower():
                        consciousness_analysis["quest_progression"].append(file_info["name"])
                        print("   🎮 Contains quest progression data")
                
                elif isinstance(content, list) and content:
                    print(f"   📋 List with {len(content)} items")
                    if content and isinstance(content[0], dict):
                        sample_keys = list(content[0].keys())[:3]
                        print(f"   🔑 Item keys: {', '.join(sample_keys)}...")
                        
            except Exception as e:
                print(f"   ⚠️ Could not analyze content: {e}")
        
        elif file_info["type"] == ".jsonl" and file_path.exists():
            try:
                with open(file_path, "r", encoding="utf-8") as f:
                    lines = f.readlines()
                print(f"   📊 JSONL with {len(lines)} entries")
                
                if lines:
                    # Parse first line to see structure
                    first_entry = json.loads(lines[0])
                    if isinstance(first_entry, dict):
                        print(f"   🔑 Entry keys: {', '.join(list(first_entry.keys())[:3])}...")
                        
            except Exception as e:
                print(f"   ⚠️ Could not analyze JSONL: {e}")
    
    # Determine consciousness health
    if consciousness_analysis["memory_patterns"] and consciousness_analysis["quest_progression"]:
        consciousness_analysis["consciousness_health"] = "active"
    elif consciousness_analysis["memory_patterns"] or consciousness_analysis["quest_progression"]:
        consciousness_analysis["consciousness_health"] = "partial"
    else:
        consciousness_analysis["consciousness_health"] = "dormant"
    
    print(f"\n🏥 Consciousness Health: {consciousness_analysis['consciousness_health'].upper()}")
    
else:
    print("🧠 No consciousness-specific data files detected")
    consciousness_analysis["consciousness_health"] = "not_detected"

analysis_data["consciousness_analysis"] = consciousness_analysis
```

## 📊 Log Analysis

```python
# Analyze logging patterns and health
log_files = (data_categories.get("json_log", []) + 
            data_categories.get("streaming_log", []) + 
            data_categories.get("text_log", []))

log_analysis = {
    "total_log_files": len(log_files),
    "total_log_size_mb": sum(f["size"] for f in log_files) / (1024 * 1024),
    "active_logs": [],
    "stale_logs": [],
    "large_logs": [],
    "log_health": "unknown"
}

print("📊 LOG ANALYSIS")
print("=" * 40)

if log_files:
    print(f"📋 Total Log Files: {len(log_files)}")
    print(f"💾 Total Log Storage: {log_analysis['total_log_size_mb']:.2f} MB")
    
    # Categorize logs by activity and size
    for log_file in log_files:
        size_mb = log_file["size"] / (1024 * 1024)
        
        # Check if log is active (modified within last 24 hours)
        if log_file["age_hours"] < 24:
            log_analysis["active_logs"].append({
                "name": log_file["name"],
                "size_mb": size_mb,
                "age_hours": log_file["age_hours"]
            })
        
        # Check if log is stale (not modified for more than 7 days)
        elif log_file["age_hours"] > 24 * 7:
            log_analysis["stale_logs"].append({
                "name": log_file["name"],
                "size_mb": size_mb,
                "age_days": log_file["age_hours"] / 24
            })
        
        # Check if log is large (> 10 MB)
        if size_mb > 10:
            log_analysis["large_logs"].append({
                "name": log_file["name"],
                "size_mb": size_mb,
                "category": log_file["category"]
            })
    
    # Display active logs
    if log_analysis["active_logs"]:
        print(f"\n🟢 ACTIVE LOGS ({len(log_analysis['active_logs'])} files):")
        for log in sorted(log_analysis["active_logs"], key=lambda x: x["age_hours"])[:5]:
            print(f"  📋 {log['name']} - {log['size_mb']:.2f} MB, {log['age_hours']:.1f}h ago")
    
    # Display stale logs
    if log_analysis["stale_logs"]:
        print(f"\n🔴 STALE LOGS ({len(log_analysis['stale_logs'])} files):")
        for log in sorted(log_analysis["stale_logs"], key=lambda x: x["age_days"], reverse=True)[:5]:
            print(f"  📋 {log['name']} - {log['size_mb']:.2f} MB, {log['age_days']:.0f} days old")
    
    # Display large logs
    if log_analysis["large_logs"]:
        print(f"\n📈 LARGE LOGS ({len(log_analysis['large_logs'])} files):")
        for log in sorted(log_analysis["large_logs"], key=lambda x: x["size_mb"], reverse=True):
            print(f"  📋 {log['name']} - {log['size_mb']:.2f} MB ({log['category']})")
    
    # Determine log health
    active_ratio = len(log_analysis["active_logs"]) / len(log_files)
    if active_ratio > 0.5:
        log_analysis["log_health"] = "healthy"
    elif active_ratio > 0.2:
        log_analysis["log_health"] = "moderate"
    else:
        log_analysis["log_health"] = "poor"
    
    print(f"\n🏥 Log Health: {log_analysis['log_health'].upper()} ({active_ratio:.1%} active)")
    
else:
    print("📊 No log files detected")
    log_analysis["log_health"] = "no_logs"

analysis_data["log_analysis"] = log_analysis
```

## 🎯 Data Management Recommendations

```python
# Generate data management recommendations
data_recommendations = []

# Storage optimization recommendations
if storage_analysis["total_size_mb"] > 100:  # More than 100 MB
    data_recommendations.append({
        "priority": "🟡 Medium",
        "category": "Storage Optimization",
        "issue": f"Data storage is {storage_analysis['total_size_mb']:.1f} MB - consider cleanup",
        "recommendation": "Implement log rotation and archive old data",
        "implementation": "Create automated cleanup scripts for logs older than 30 days"
    })

# Log management recommendations
if log_analysis["log_health"] in ["poor", "moderate"]:
    data_recommendations.append({
        "priority": "🔴 High" if log_analysis["log_health"] == "poor" else "🟡 Medium",
        "category": "Log Management",
        "issue": f"Log health is {log_analysis['log_health']} - many stale logs detected",
        "recommendation": "Implement active logging and cleanup stale logs",
        "implementation": "Review logging configuration and remove unused log files"
    })

if log_analysis["large_logs"]:
    data_recommendations.append({
        "priority": "🟡 Medium",
        "category": "Large File Management",
        "issue": f"{len(log_analysis['large_logs'])} large log files detected",
        "recommendation": "Implement log rotation for large files",
        "implementation": "Set up automatic log rotation at 10MB threshold",
        "affected_files": [log["name"] for log in log_analysis["large_logs"][:3]]
    })

# Consciousness data recommendations
if consciousness_analysis["consciousness_health"] == "dormant":
    data_recommendations.append({
        "priority": "🔵 Enhancement",
        "category": "Consciousness Integration",
        "issue": "Consciousness data appears dormant",
        "recommendation": "Activate consciousness bridge data collection",
        "implementation": "Initialize consciousness memory systems and quest tracking"
    })
elif consciousness_analysis["consciousness_health"] == "partial":
    data_recommendations.append({
        "priority": "🟢 Low",
        "category": "Consciousness Enhancement",
        "issue": "Partial consciousness data detected",
        "recommendation": "Complete consciousness data integration",
        "implementation": "Ensure all consciousness systems are actively logging"
    })

# Backup recommendations
consciousness_size = consciousness_analysis["total_consciousness_size"] / 1024
if consciousness_size > 10:  # More than 10 KB of consciousness data
    data_recommendations.append({
        "priority": "🟢 Low",
        "category": "Data Backup",
        "issue": f"{consciousness_size:.1f} KB of consciousness data needs protection",
        "recommendation": "Implement regular backup of consciousness data",
        "implementation": "Create automated backup of critical consciousness files"
    })

# Data organization recommendations
if len(analysis_data["file_types"]) > 8:
    data_recommendations.append({
        "priority": "🟢 Low",
        "category": "Data Organization",
        "issue": f"{len(analysis_data['file_types'])} different file types - consider standardization",
        "recommendation": "Standardize on fewer data formats",
        "implementation": "Prefer JSON for structured data, JSONL for logs, minimize other formats"
    })

print("🎯 DATA MANAGEMENT RECOMMENDATIONS")
print("=" * 50)

for i, rec in enumerate(data_recommendations, 1):
    print(f"{i}. {rec['priority']} - {rec['category']}")
    print(f"   📋 Issue: {rec['issue']}")
    print(f"   💡 Recommendation: {rec['recommendation']}")
    print(f"   🔧 Implementation: {rec['implementation']}")
    if "affected_files" in rec and rec["affected_files"]:
        print(f"   📁 Affected Files: {', '.join(rec['affected_files'])}")
    print()

analysis_data["data_recommendations"] = data_recommendations

# Generate final statistics
data_stats = {
    "total_files": storage_analysis["total_files"],
    "total_size_mb": storage_analysis["total_size_mb"],
    "subdirectories": len(analysis_data["subdirectories"]),
    "file_types": len(analysis_data["file_types"]),
    "log_files": log_analysis["total_log_files"],
    "consciousness_files": consciousness_analysis["consciousness_files"] + consciousness_analysis["quest_files"],
    "active_logs": len(log_analysis["active_logs"]),
    "consciousness_health": consciousness_analysis["consciousness_health"],
    "log_health": log_analysis["log_health"]
}

print("📊 DATA MANAGEMENT STATISTICS SUMMARY")
print("=" * 40)
for key, value in data_stats.items():
    print(f"📈 {key.replace('_', ' ').title()}: {value}")

analysis_data["data_stats"] = data_stats
```
