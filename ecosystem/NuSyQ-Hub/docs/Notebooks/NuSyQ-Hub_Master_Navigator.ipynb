```python
# 🏥 Repository Health Check - LIVE INTEGRATION
# Connects to real ZETA tracker and system metrics

import json
import sys
from pathlib import Path

# Repository root path
REPO_ROOT = Path.cwd().parent.parent
sys.path.insert(0, str(REPO_ROOT))

def load_zeta_tracker():
    """Load live ZETA progress data"""
    zeta_path = REPO_ROOT / "config" / "ZETA_PROGRESS_TRACKER.json"
    if zeta_path.exists():
        with open(zeta_path, "r", encoding="utf-8") as f:
            return json.load(f)
    return None

def _classify_task_status(status, state):
    """Classify a task into a status category"""
    if status == "●" or state == "MASTERED":
        return "mastered"
    elif status in ["◐", "◑"] or state == "IN-PROGRESS":
        return "in_progress"
    elif state == "INITIALIZED":
        return "initialized"
    return None

def _count_task_statuses(phases):
    """Count tasks by status from phases data"""
    counts = {"mastered": 0, "in_progress": 0, "initialized": 0, "total": 0}
    for phase_data in phases.values():
        for task in phase_data.get("tasks", []):
            counts["total"] += 1
            status_key = _classify_task_status(
                task.get("status", "○"),
                task.get("state", "INITIALIZED")
            )
            if status_key:
                counts[status_key] += 1
    return counts

def analyze_zeta_status(zeta_data):
    """Analyze current ZETA progress"""
    if not zeta_data:
        return {"status": "ZETA tracker not found"}
    
    progress = zeta_data.get("current_progress", {})
    counts = _count_task_statuses(zeta_data.get("phases", {}))
    
    return {
        "current_phase": progress.get("current_phase"),
        "completion": f"{progress.get('completion_percentage', 0)}%",
        "mastery": f"{progress.get('mastery_percentage', 0)}%",
        "mastered_count": counts["mastered"],
        "in_progress_count": counts["in_progress"],
        "available_count": counts["initialized"],
        "total_tasks": counts["total"],
        "recent_achievements": progress.get("recent_achievements", [])[:5]
    }

def basic_health_check():
    """Live repository health validation"""
    health_status = {
        "repository_root": str(REPO_ROOT),
        "core_directories": {},
        "key_files": {},
        "zeta_status": None,
        "system_status": "OPERATIONAL"
    }
    
    # Check core directories
    core_dirs = ["src", "config", "docs", "data", "tests"]
    for dir_name in core_dirs:
        dir_path = REPO_ROOT / dir_name
        health_status["core_directories"][dir_name] = {
            "exists": dir_path.exists(),
            "path": str(dir_path)
        }
    
    # Check key configuration files
    key_files = [
        "config/ZETA_PROGRESS_TRACKER.json",
        "pyproject.toml",
        "README.md"
    ]
    for file_rel in key_files:
        file_path = REPO_ROOT / file_rel
        health_status["key_files"][file_rel] = file_path.exists()
    
    # Load ZETA tracker data
    zeta_data = load_zeta_tracker()
    health_status["zeta_status"] = analyze_zeta_status(zeta_data)
    
    return health_status

# Execute health check
print("🔍 Running Live Repository Health Check...")
health = basic_health_check()

print(f"\n📍 Repository: {health['repository_root']}")
print(f"✅ System Status: {health['system_status']}")

print("\n📂 Core Directories:")
for dir_name, info in health["core_directories"].items():
    status = "✅" if info["exists"] else "❌"
    print(f"  {status} {dir_name}/")

print("\n📄 Key Configuration Files:")
for file_name, exists in health["key_files"].items():
    status = "✅" if exists else "❌"
    print(f"  {status} {file_name}")

if health["zeta_status"]:
    zeta = health["zeta_status"]
    print("\n🎯 ZETA Progress Tracker (LIVE):")
    print(f"  📊 Completion: {zeta['completion']} | Mastery: {zeta['mastery']}")
    print(f"  ● Mastered: {zeta['mastered_count']}")
    print(f"  ◐ In Progress: {zeta['in_progress_count']}")
    print(f"  ○ Available: {zeta['available_count']}")
    print(f"  📋 Total Tasks: {zeta['total_tasks']}")
    print(f"\n  🏆 Current Phase: {zeta['current_phase']}")
    
    if zeta.get("recent_achievements"):
        print("\n  ⚡ Recent Achievements:")
        for achievement in zeta["recent_achievements"]:
            print(f"    • {achievement}")

print("\n✅ Health check complete!")
```

```python
try:
    import ipywidgets as widgets
    from IPython.display import HTML, display
    WIDGETS_AVAILABLE = True
except ImportError:
    WIDGETS_AVAILABLE = False
    print("⚠️ ipywidgets not available - install with: pip install ipywidgets")

def create_quest_browser():
    """Create interactive quest browser"""
    if not WIDGETS_AVAILABLE:
        return
    
    zeta_data = load_zeta_tracker()
    if not zeta_data:
        print("❌ ZETA tracker not available")
        return
    
    # Build quest list
    all_quests = []
    for phase_data in zeta_data.get("phases", {}).values():
        for task in phase_data.get("tasks", []):
            all_quests.append({
                "id": task.get("id", "Unknown"),
                "description": task.get("description", "No description"),
                "status": task.get("status", "○"),
                "state": task.get("state", "INITIALIZED"),
                "phase": phase_data.get("name", "Unknown"),
                "progress_note": task.get("progress_note", "")
            })
    
    # Create filter dropdown
    filter_options = ["All", "Mastered (●)", "In Progress (◐)", "Available (○)"]
    filter_dropdown = widgets.Dropdown(
        options=filter_options,
        value="All",
        description="Filter:",
        style={"description_width": "initial"}
    )
    
    # Create output area
    output = widgets.Output()
    
    def update_quest_list(change):
        with output:
            output.clear_output()
            
            filter_val = change["new"]
            
            # Filter quests
            if filter_val == "Mastered (●)":
                filtered = [q for q in all_quests if q["status"] == "●" or q["state"] == "MASTERED"]
            elif filter_val == "In Progress (◐)":
                filtered = [q for q in all_quests if q["status"] in ["◐", "◑"] or q["state"] == "IN-PROGRESS"]
            elif filter_val == "Available (○)":
                filtered = [q for q in all_quests if q["status"] == "○" and q["state"] == "INITIALIZED"]
            else:
                filtered = all_quests
            
            # Display quests
            html = "<div style='font-family: monospace;'>"
            html += f"<h3>📋 ZETA Quests ({len(filtered)} shown)</h3>"
            
            for quest in filtered:
                status_emoji = quest["status"]
                state_color = {
                    "MASTERED": "#28a745",
                    "IN-PROGRESS": "#ffc107",
                    "INITIALIZED": "#6c757d"
                }.get(quest["state"], "#6c757d")
                
                html += f"<div style='margin: 10px 0; padding: 10px; border-left: 4px solid {state_color}; background: #f8f9fa;'>"
                html += f"<strong>{status_emoji} {quest['id']}</strong> - {quest['description']}<br>"
                html += f"<small>Phase: {quest['phase']} | State: {quest['state']}</small>"
                if quest["progress_note"]:
                    html += f"<br><em style='color: #007bff;'>Progress: {quest['progress_note']}</em>"
                html += "</div>"
            
            html += "</div>"
            display(HTML(html))
    
    filter_dropdown.observe(update_quest_list, names="value")
    
    # Initial display
    display(filter_dropdown)
    display(output)
    
    # Trigger initial update
    update_quest_list({"new": "All"})

# Create and display quest browser
print("🎮 Initializing Interactive Quest Browser...")
create_quest_browser()
```
