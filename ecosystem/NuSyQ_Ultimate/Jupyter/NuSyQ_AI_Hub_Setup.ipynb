## 6. 🤖 AI Integration Setup

Now let's set up integration with your Ollama models and create helper functions for AI development workflows.

````python
# AI Integration and Ollama Model Management
import requests
import time
from datetime import datetime

class NuSyQ_AI_Hub:
    """Main class for managing AI models and workflows in the NuSyQ environment"""
    
    def __init__(self):
        self.ollama_base_url = "http://localhost:11434"
        self.models_from_manifest = [
            "qwen2.5-coder:7b", "qwen2.5-coder:14b", "codellama:7b",
            "deepseek-coder:v2", "deepseek-coder-v3:35b-q4_K_M",
            "aixcoder-7b", "phi4", "starcoder2:15b", "gemma2:9b", "phi3.5"
        ]
        
    def check_ollama_status(self):
        """Check if Ollama is running and accessible"""
        try:
            response = requests.get(f"{self.ollama_base_url}/api/tags", timeout=5)
            if response.status_code == 200:
                return True, "Ollama is running"
            else:
                return False, f"Ollama responded with status {response.status_code}"
        except requests.exceptions.RequestException as e:
            return False, f"Cannot connect to Ollama: {e}"
    
    def list_available_models(self):
        """List all models currently available in Ollama"""
        try:
            response = requests.get(f"{self.ollama_base_url}/api/tags", timeout=10)
            if response.status_code == 200:
                models = response.json()
                return True, models.get('models', [])
            else:
                return False, f"Failed to get models: {response.status_code}"
        except Exception as e:
            return False, f"Error getting models: {e}"
    
    def check_model_status(self):
        """Check which models from manifest are installed"""
        success, models = self.list_available_models()
        if not success:
            return {"error": models}
        
        installed_models = [model['name'] for model in models]
        status = {}
        
        for model in self.models_from_manifest:
            status[model] = {
                'installed': any(model in installed for installed in installed_models),
                'full_name': next((m for m in installed_models if model in m), None)
            }
        
        return status
    
    def send_prompt(self, model, prompt, stream=False):
        """Send a prompt to an Ollama model"""
        try:
            payload = {
                "model": model,
                "prompt": prompt,
                "stream": stream
            }
            
            response = requests.post(
                f"{self.ollama_base_url}/api/generate",
                json=payload,
                timeout=30
            )
            
            if response.status_code == 200:
                return True, response.json()
            else:
                return False, f"Request failed: {response.status_code}"
        except Exception as e:
            return False, f"Error sending prompt: {e}"

# Initialize the AI Hub
ai_hub = NuSyQ_AI_Hub()

# Check Ollama status
print("🤖 NuSyQ AI Hub Initialization")
print("=" * 40)

ollama_running, status_msg = ai_hub.check_ollama_status()
print(f"Ollama Status: {'✅ Running' if ollama_running else '❌ Not Available'}")
print(f"Details: {status_msg}")

if ollama_running:
    print(f"\n📊 Model Status Check:")
    model_status = ai_hub.check_model_status()
    
    if 'error' in model_status:
        print(f"❌ Error checking models: {model_status['error']}")
    else:
        installed_count = sum(1 for status in model_status.values() if status['installed'])
        total_count = len(model_status)
        
        print(f"Models installed: {installed_count}/{total_count}")
        
        for model, status in model_status.items():
            status_icon = "✅" if status['installed'] else "⏳"
            full_name = status['full_name'] if status['installed'] else "Not installed"
            print(f"  {status_icon} {model}: {full_name}")
        
        if installed_count > 0:
            print(f"\n🎉 Great! You have {installed_count} models ready for AI development!")
        else:
            print(f"\n⏳ Models are still downloading. Check back later!")
            
else:
    print(f"\n⚠️ Ollama is not running. Start Ollama to use AI models.")

print(f"\n🔗 AI Hub class created as 'ai_hub' - use it to interact with your models!")
````

## 7. 📊 System Monitoring Dashboard

Let's create a system monitoring dashboard to track resources and model performance.

````python
# System Monitoring Dashboard
import psutil
import platform
from datetime import datetime
from IPython.display import display, HTML

def create_system_dashboard():
    """Create a comprehensive system monitoring dashboard"""
    
    # System Information
    system_info = {
        "System": platform.system(),
        "Node": platform.node(),
        "Release": platform.release(),
        "Version": platform.version(),
        "Machine": platform.machine(),
        "Processor": platform.processor()
    }
    
    # Memory Information
    memory = psutil.virtual_memory()
    memory_info = {
        "Total": f"{memory.total / (1024**3):.1f} GB",
        "Available": f"{memory.available / (1024**3):.1f} GB",
        "Used": f"{memory.used / (1024**3):.1f} GB",
        "Percentage": f"{memory.percent}%"
    }
    
    # CPU Information
    cpu_info = {
        "Physical cores": psutil.cpu_count(logical=False),
        "Total cores": psutil.cpu_count(logical=True),
        "Max Frequency": f"{psutil.cpu_freq().max:.0f} MHz" if psutil.cpu_freq() else "Unknown",
        "Current Frequency": f"{psutil.cpu_freq().current:.0f} MHz" if psutil.cpu_freq() else "Unknown",
        "CPU Usage": f"{psutil.cpu_percent(interval=1)}%"
    }
    
    # Disk Information  
    disk_info = {}
    partitions = psutil.disk_partitions()
    for partition in partitions:
        try:
            disk_usage = psutil.disk_usage(partition.mountpoint)
            disk_info[partition.device] = {
                "Total": f"{disk_usage.total / (1024**3):.1f} GB",
                "Used": f"{disk_usage.used / (1024**3):.1f} GB",
                "Free": f"{disk_usage.free / (1024**3):.1f} GB",
                "Percentage": f"{(disk_usage.used / disk_usage.total) * 100:.1f}%"
            }
        except PermissionError:
            continue
    
    # Create HTML dashboard
    html = f"""
    <div style="background: linear-gradient(135deg, #667eea 0%, #764ba2 100%); 
                color: white; padding: 20px; border-radius: 10px; font-family: 'Segoe UI', Arial, sans-serif;">
        <h2 style="margin-top: 0; text-align: center;">🖥️ NuSyQ System Dashboard</h2>
        <p style="text-align: center; margin-bottom: 30px;"><strong>Generated:</strong> {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}</p>
        
        <div style="display: grid; grid-template-columns: repeat(auto-fit, minmax(300px, 1fr)); gap: 20px;">
            
            <!-- System Info -->
            <div style="background: rgba(255,255,255,0.1); padding: 15px; border-radius: 8px;">
                <h3 style="margin-top: 0; color: #ffeb3b;">🖥️ System Information</h3>
                <ul style="list-style: none; padding: 0;">
                    {''.join([f'<li><strong>{k}:</strong> {v}</li>' for k, v in system_info.items()])}
                </ul>
            </div>
            
            <!-- Memory Info -->
            <div style="background: rgba(255,255,255,0.1); padding: 15px; border-radius: 8px;">
                <h3 style="margin-top: 0; color: #4caf50;">🧠 Memory Status</h3>
                <ul style="list-style: none; padding: 0;">
                    {''.join([f'<li><strong>{k}:</strong> {v}</li>' for k, v in memory_info.items()])}
                </ul>
            </div>
            
            <!-- CPU Info -->
            <div style="background: rgba(255,255,255,0.1); padding: 15px; border-radius: 8px;">
                <h3 style="margin-top: 0; color: #ff9800;">⚡ CPU Information</h3>
                <ul style="list-style: none; padding: 0;">
                    {''.join([f'<li><strong>{k}:</strong> {v}</li>' for k, v in cpu_info.items()])}
                </ul>
            </div>
            
            <!-- Disk Info -->
            <div style="background: rgba(255,255,255,0.1); padding: 15px; border-radius: 8px;">
                <h3 style="margin-top: 0; color: #2196f3;">💾 Disk Usage</h3>
                {''.join([f'''
                <div style="margin-bottom: 10px;">
                    <strong>{device}</strong><br>
                    <small>Total: {info["Total"]} | Used: {info["Used"]} | Free: {info["Free"]} ({info["Percentage"]})</small>
                </div>
                ''' for device, info in disk_info.items()])}
            </div>
        </div>
        
        <div style="margin-top: 20px; text-align: center; opacity: 0.8;">
            <small>💡 This dashboard updates each time you run this cell</small>
        </div>
    </div>
    """
    
    return html

# Generate and display the dashboard
dashboard_html = create_system_dashboard()
display(HTML(dashboard_html))

# Also provide a quick summary for the console
print("📊 System Quick Summary:")
print(f"  💾 Memory: {psutil.virtual_memory().available / (1024**3):.1f} GB available")
print(f"  ⚡ CPU: {psutil.cpu_percent(interval=1)}% usage")
print(f"  🖥️ Platform: {platform.system()} {platform.release()}")
print(f"  🕒 Uptime: {datetime.now().strftime('%Y-%m-%d %H:%M:%S')}")

# Check if we're ready for AI workloads
memory_gb = psutil.virtual_memory().total / (1024**3)
cpu_cores = psutil.cpu_count(logical=True)

print(f"\n🤖 AI Readiness Assessment:")
if memory_gb >= 16 and cpu_cores >= 8:
    print("✅ Excellent! Your system is well-equipped for AI development")
    if memory_gb >= 32:
        print("🚀 You can run large models (14B+ parameters) smoothly")
    else:
        print("👍 You can run medium models (7B-14B parameters) effectively")
else:
    print("⚠️ Limited resources - consider smaller models or cloud computing")
````

## 📊 Real-Time Model Download Progress

Let's monitor our model downloads and continue pulling the remaining models from the manifest.

````python
# Model Download Progress Tracker
import subprocess
import time
from datetime import datetime

def get_download_status():
    """Check current model download status"""
    target_models_from_manifest = [
        "qwen2.5-coder:7b", "qwen2.5-coder:14b", "codellama:7b",
        "deepseek-coder:v2", "deepseek-coder-v3:35b-q4_K_M",
        "aixcoder-7b", "phi4", "starcoder2:15b", "gemma2:9b", "phi3.5"
    ]
    
    currently_downloading = [
        "qwen2.5-coder:14b", "codellama:7b", "phi3.5", "gemma2:9b", "deepseek-coder"
    ]
    
    # Get current ollama list
    try:
        result = subprocess.run(['ollama', 'list'], capture_output=True, text=True, timeout=10)
        if result.returncode == 0:
            lines = result.stdout.strip().split('\n')
            installed = []
            if len(lines) > 1:  # Skip header
                for line in lines[1:]:
                    if line.strip():
                        model_name = line.split()[0]
                        installed.append(model_name)
            
            print(f"🤖 NuSyQ Model Download Status - {datetime.now().strftime('%H:%M:%S')}")
            print("=" * 60)
            print(f"📊 Progress: {len(installed)} installed, {len(currently_downloading)} downloading")
            print()
            
            # Status for each model
            for model in target_models_from_manifest:
                if model in installed:
                    print(f"✅ {model} - READY")
                elif any(model.split(':')[0] in dl.split(':')[0] for dl in currently_downloading):
                    print(f"⏳ {model} - DOWNLOADING")
                else:
                    print(f"⏸️ {model} - QUEUED")
            
            print()
            print(f"🚀 Currently downloading: {', '.join(currently_downloading)}")
            return len(installed), len(target_models_from_manifest)
        else:
            print("❌ Could not connect to Ollama")
            return 0, len(target_models_from_manifest)
    except subprocess.TimeoutExpired:
        print("⏳ Ollama is busy (models downloading)")
        return 0, len(target_models_from_manifest)
    except Exception as e:
        print(f"❌ Error: {e}")
        return 0, len(target_models_from_manifest)

# Check current status
ready_count, total_count = get_download_status()

print(f"\n🎯 Next Steps:")
print("1. Keep this terminal running to continue downloads")
print("2. Run this cell periodically to check progress") 
print("3. Once models are ready, test them with ai_hub.send_prompt()")
print(f"\n💡 Tip: Open the Model_Download_Monitor.ipynb for automated tracking!")
````

## 🧪 Practical AI Testing (Quick Version)

Let's test our available model with a working configuration:

````python
# Quick AI Test with Optimized Settings
import requests
import json
import time

class QuickAI:
    def __init__(self):
        self.base_url = "http://localhost:11434"
    
    def quick_prompt(self, model, prompt, max_tokens=100):
        """Send a prompt with optimized settings for quick response"""
        try:
            payload = {
                "model": model,
                "prompt": prompt,
                "stream": False,
                "options": {
                    "num_predict": max_tokens,
                    "temperature": 0.7,
                    "top_p": 0.9
                }
            }
            
            print(f"🤖 Asking {model}: {prompt[:50]}...")
            start_time = time.time()
            
            response = requests.post(
                f"{self.base_url}/api/generate",
                json=payload,
                timeout=60  # Increased timeout
            )
            
            end_time = time.time()
            
            if response.status_code == 200:
                result = response.json()
                response_text = result.get('response', 'No response')
                print(f"⚡ Response time: {end_time - start_time:.1f}s")
                print(f"💬 Response: {response_text}")
                return True, response_text
            else:
                print(f"❌ HTTP Error: {response.status_code}")
                return False, f"HTTP {response.status_code}"
                
        except requests.exceptions.Timeout:
            print("⏳ Request timed out - model might be busy with downloads")
            return False, "Timeout"
        except Exception as e:
            print(f"❌ Error: {e}")
            return False, str(e)
    
    def test_available_models(self):
        """Test all available models with quick prompts"""
        try:
            # Get available models
            response = requests.get(f"{self.base_url}/api/tags", timeout=10)
            if response.status_code == 200:
                models = response.json().get('models', [])
                model_names = [m['name'] for m in models]
                
                print(f"🎯 Testing {len(model_names)} available models...")
                print("=" * 50)
                
                test_prompt = "Write a simple Python function to add two numbers"
                
                for model_name in model_names:
                    print(f"\n🔸 Testing {model_name}:")
                    success, result = self.quick_prompt(model_name, test_prompt, max_tokens=50)
                    if not success:
                        print(f"   Skipping {model_name} - {result}")
                    print("-" * 30)
                
                return True
            else:
                print("❌ Could not get model list")
                return False
        except Exception as e:
            print(f"❌ Error testing models: {e}")
            return False

# Initialize and test
print("🚀 Quick AI Testing System")
print("=" * 40)

quick_ai = QuickAI()

# Test connection first
try:
    response = requests.get("http://localhost:11434/api/tags", timeout=5)
    if response.status_code == 200:
        print("✅ Ollama connection successful")
        quick_ai.test_available_models()
    else:
        print(f"❌ Ollama not responding properly: {response.status_code}")
except Exception as e:
    print(f"❌ Cannot connect to Ollama: {e}")
    print("💡 Make sure Ollama is running and not busy with downloads")

print(f"\n🎯 Quick AI testing complete!")
````
