✅ llama3.1:8b - General AI assistant

```python
# Quick AI Development Setup
import requests
import json

class DevAI:
    def __init__(self):
        self.base_url = "http://localhost:11434"
        self.coder = "qwen2.5-coder:7b"
        self.general = "llama3.1:8b"
    
    def code_assist(self, prompt, max_tokens=200):
        """Get coding help from the specialist model"""
        return self._ask(self.coder, prompt, max_tokens)
    
    def general_help(self, prompt, max_tokens=200):
        """Get general assistance"""
        return self._ask(self.general, prompt, max_tokens)
    
    def _ask(self, model, prompt, max_tokens):
        try:
            response = requests.post(f"{self.base_url}/api/generate", 
                json={
                    "model": model,
                    "prompt": prompt,
                    "stream": False,
                    "options": {"num_predict": max_tokens}
                }, timeout=60)
            
            if response.status_code == 200:
                return response.json()['response']
            else:
                return f"Error: {response.status_code}"
        except Exception as e:
            return f"Error: {e}"

# Initialize your AI assistant
ai = DevAI()
print("🤖 AI Development Assistant Ready!")
print("Use: ai.code_assist('your coding question')")
print("Use: ai.general_help('your general question')")
```
