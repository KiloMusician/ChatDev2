# 🚀 Quick AI Model Test

Use this notebook to quickly test your AI models once they're downloaded.

```python
# Quick setup - import the AI Hub
import sys
sys.path.append(r'c:\Users\keath\NuSyQ\Jupyter')

# If you've run the setup notebook, this class will be available
import requests

class QuickAI:
    def __init__(self):
        self.base_url = "http://localhost:11434"
    
    def quick_test(self, model="qwen2.5-coder:7b", prompt="Hello! Can you help me with Python?"):
        try:
            response = requests.post(f"{self.base_url}/api/generate", 
                                   json={"model": model, "prompt": prompt, "stream": False})
            if response.status_code == 200:
                return response.json()['response']
            else:
                return f"Error: {response.status_code}"
        except Exception as e:
            return f"Connection error: {e}"

ai = QuickAI()
print("🤖 Quick AI Test initialized!")
print("Use: ai.quick_test('model-name', 'your prompt')")
```
