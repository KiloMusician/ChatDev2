# 🎭 AI Group Chat & Integration Hub

**Multi-AI system testing and integration**

Available AI Systems:
- 🤖 **Ollama Models** (qwen2.5-coder:7b, llama3.1:8b)
- 🔮 **Claude Code** (VS Code extension)
- 🐙 **GitHub Copilot** (Code completion + chat)
- ⚡ **Kilo Code** (Additional AI assistance)

```python
# AI Integration Testing Framework
import requests
import subprocess
import json
import time
from datetime import datetime

class AIEcosystem:
    def __init__(self, ollama_url="http://localhost:11434", max_tokens=100):
        self.ollama_url = ollama_url
        self.max_tokens = max_tokens
        self.available_models = []
        self.test_results = {}
        
    def check_ollama_models(self):
        """Check available Ollama models"""
        try:
            response = requests.get(f"{self.ollama_url}/api/tags", timeout=5)
            if response.status_code == 200:
                models = response.json().get('models', [])
                self.available_models = [m['name'] for m in models]
                return True, self.available_models
            return False, []
        except Exception as e:
            return False, str(e)
    
    def check_vscode_extensions(self):
        """Check VS Code AI extensions"""
        try:
            result = subprocess.run(['code', '--list-extensions'], 
                                  capture_output=True, text=True, timeout=10)
            if result.returncode == 0:
                extensions = result.stdout.strip().split('\n')
                ai_extensions = [ext for ext in extensions if any(keyword in ext.lower() 
                    for keyword in ['anthropic', 'claude', 'copilot', 'kilo', 'ai'])]
                return True, ai_extensions
            return False, []
        except Exception as e:
            return False, str(e)
    
    def test_ai_conversation(self, model, prompt):
        """Test AI model with a prompt"""
        try:
            start_time = time.time()
            response = requests.post(f"{self.ollama_url}/api/generate",
                json={
                    "model": model,
                    "prompt": prompt,
                    "stream": False,
                    "options": {"num_predict": self.max_tokens}
                }, timeout=30)
            
            if response.status_code == 200:
                result = response.json()
                response_time = time.time() - start_time
                return {
                    "success": True,
                    "response": result.get('response', ''),
                    "time": response_time,
                    "model": model
                }
            else:
                return {"success": False, "error": f"HTTP {response.status_code}"}
        except Exception as e:
            return {"success": False, "error": str(e)}
    
    def ai_group_chat_simulation(self):
        """Simulate a group chat between different AI models"""
        conversation_topic = "Best practices for AI-assisted software development"
        
        print(f"🎭 AI Group Chat Simulation: {conversation_topic}")
        print("=" * 60)
        
        # Check available models
        success, models = self.check_ollama_models()
        if not success:
            print(f"❌ Cannot connect to Ollama: {models}")
            return
        
        print(f"👥 Participants: {', '.join(models)}")
        print()
        
        # Simulate conversation
        prompts = [
            "What are the key benefits of AI-assisted coding?",
            "How can developers maintain code quality when using AI tools?",
            "What's your perspective on the previous responses?"
        ]
        
        for i, prompt in enumerate(prompts, 1):
            print(f"🗨️ Round {i}: {prompt}")
            print("-" * 40)
            
            for model in models:
                result = self.test_ai_conversation(model, prompt)
                if result['success']:
                    print(f"🤖 {model} ({result['time']:.1f}s):")
                    print(f"   {result['response'][:200]}...")
                else:
                    print(f"❌ {model}: {result['error']}")
                print()
            
            time.sleep(1)  # Brief pause between rounds
        
        print("🎉 Group chat simulation complete!")

# Initialize AI Ecosystem
ai_ecosystem = AIEcosystem()
print("🚀 AI Ecosystem initialized!")
```

## 🔍 System Status Check

```python
# Check all AI systems
print("🔍 AI Ecosystem Status Check")
print("=" * 40)

# Check Ollama models
print("🤖 Ollama Models:")
success, models = ai_ecosystem.check_ollama_models()
if success:
    for model in models:
        print(f"  ✅ {model}")
else:
    print(f"  ❌ {models}")

print()

# Check VS Code extensions
print("🔌 VS Code AI Extensions:")
success, extensions = ai_ecosystem.check_vscode_extensions()
if success:
    for ext in extensions:
        print(f"  ✅ {ext}")
else:
    print(f"  ❌ {extensions}")

print()
print("📊 Integration Status: Ready for multi-AI workflows!")
```
